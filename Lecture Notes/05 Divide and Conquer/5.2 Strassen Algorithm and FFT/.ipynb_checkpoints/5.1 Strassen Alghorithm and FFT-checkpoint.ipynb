<br>
در این بخش چگونگی ضرب دو ماتریس را مورد بررسی قرار می‌دهیم و به دنبال راه‌حلی سریع و بهینه برای این کار هستیم.
<br>
    
 فرض کنید دو ماتریس 
$A_{n*n}$
و
$B_{n*n}$
را

<h2 style="color: #ff6600;">
راه حل اول(راه حل بدیهی)
</h2>         
طبق تعریف ضرب ماتریس‌ها عمل کرده و هر سطر از ماتریس 
$A$
را در هر ستون از ماتریس 
$B$
ضرب می‌کنیم.
<img src="matrix_multiplication.png" style="width:60%">
<strong>
    پیچیدگی زمانی:
</strong>
برای محاسبه‌ی هر درایه از ماتریس
$C$
باید ضرب یک سطر در یک ستون انجام شود که از مرتبه زمانی 
$O(n)$
زمان می‌برد.
و چون ماتریس 
$C$
دارای
$n^2$
درایه است، انجام ضرب با این روش در مرتبه‌ی زمانی 
$O(n^3)$
انجام می‌شود.
<hr>
</div>

```python
int ** regularMultiple(int ** a , int ** b, int size){

    int** c = new int*[size];
    for (int i = 0; i < size; i++)
        c[i] = new int[size];

    for (int i = 0; i < size; i++) {
        for (int j = 0; j < size; j++) {
            c[i][j] = 0;
            for (int k = 0; k < size; k++) {
                c[i][j] += a[i][k] * b[k][j];
            }
        }
    }
    return c;
}
```

```python
void printMatrix(int ** A, int size)
{
    for (int i = 0; i < size; i++)
    {
        for (int j = 0; j < size; j++)
        {
            cout << A[i][j] << " ";
        }
        cout << "\n";
    }
}
```

```python
{
    //define first matrix
    int** A;
    A = new int*[2];
    for (int i = 0; i < 2; i++)
        A[i] = new int[2];
    A[0][0] = 1;
    A[0][1] = 2;
    A[1][0] = 4;
    A[1][1] = 5;
    
    cout << "first matrix is: \n";
    printMatrix(A, 2);
    cout << "\n";
    
    

    //define second matrix
    int ** B;
    B = new int*[2];
    for (int i = 0; i < 3; i++)
        B[i] = new int[2];
    B[0][0] = 1;
    B[0][1] = 2;
    B[1][0] = 1;
    B[1][1] = 2;
    
    cout << "second matrix is: \n";
    printMatrix(B, 2);
    cout << "\n";

    
    //compute answer
    int** C = regularMultiple(A, B, 2);
    
    //print answer
    cout << "the answer is: \n";
    printMatrix(C, 2);
    
   

}
```

<h2 style="color: #ff6600;">
راه حل دوم (الگوریتم استراسن)
</h2>            
ایده‌ی اصلی این الگوریتم تقسیم و حل است. دو ماتریس با ابعاد 
$n \times n$
$A$
و
$B$
داده شده است. با فرض اینکه عدد 
$n$
به توانی از عدد
$2$
است.
$(n = 2 ^ k   k \in N )$
هر کدام از دو ماتریس
$A$
و
$B$
را به 4 زیرماتریس تقسیم می‌کنیم که هر کدام در ابعاد $\frac{n}{2} \times \frac{n}{2} $
هستند.
(هر کدام از 
$A_{ij}$
ها و
$B_{ij}$
ها ماتریس هستند.)
<br/>


\begin{equation}
A=
\begin{pmatrix}
A_{11} & A_{12} \\
A_{21} & A_{22}
\end{pmatrix}
,
B=
\begin{pmatrix}
B_{11} & B_{12} \\
B_{21} & B_{22}
\end{pmatrix}
\end{equation}


<hr>
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
</div>            
به راحتی می‌توان اثبات کرد:
\begin{equation}
C=
\begin{pmatrix}
C_{11} & C_{12} \\
C_{21} & C_{22}
\end{pmatrix}
=
\begin{pmatrix}
A_{11}B_{11} + A_{12}B_{21} & A_{11}B_{12} + A_{12}B_{22} \\
A_{21}B_{11} + A_{22}B_{21} & A_{21}B_{12} + A_{22}B_{22}
\end{pmatrix}
\end{equation}
<img src="strassen_new.png" style="width:50%">

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
در اینجا در هر مرحله باید 8 ضرب ماتریس‌های 
$\frac{n}{2} * \frac{n}{2} $
را حساب کنیم, یعنی:
\begin{equation}
T(n) = 8T(\frac{n}{2}) + O(n^2)
\end{equation}
که با محاسبه با قضیه اصلی بدست می‌آید:
$T(n) \in O(n^3)$
<br/>
می‌بینیم که زمان اجرای ضرب هنوز بهبود پیدا نکرده است. ایده‌ی اشتراسن برای محاسبه‌ی ضرب ماتریس‌ها این بود که به‌جای محاسبه‌ی این 8 عبارت  
7 عبارت جدید را محاسبه کند. این 7 عبارت باید به صورتی می‌بودند که بتوان به راحتی با استفاده از آن‌ها حاصل زیر‌ماتریس‌های $C$ را محاسبه کرد. 

<br/>
این 7 عبارت به شرح زیر است:
\begin{equation}
M_1 = (A_{11} + A_{22})(B_{11} + B_{22})
\end{equation}
\begin{equation}
M_2 = (A_{21} + A_{22})(B_{11})
\end{equation}
\begin{equation}
M_3 = (A_{11})(B_{12} - B_{22})
\end{equation}
\begin{equation}
M_4 = (A_{22})(B_{21} - B_{11})
\end{equation}
\begin{equation}
M_5 = (A_{11} + A_{12})(B_{22})
\end{equation}
\begin{equation}
M_6 = (A_{21} - A_{11})(B_{11} + B_{12})
\end{equation}
\begin{equation}
M_7 = (A_{12} - A_{22})(B_{21} + B_{22})
\end{equation}
با کمی محاسبه می‌توانید مشاهده کنید که زیرماتریس‌های زیر به شکل زیر بدست می‌آیند.
\begin{equation}
C_{11} = M_1 + M_4 + M_7 - M_5 \\
C_{12} = M_3 + M_5 \\
C_{21} = M_2 + M_4 \\ 
C_{22} = M_1 + M_3 + M_6 - M_2
\end{equation}

<strong>
    نکته:
</strong>
در ابتدای این روش فرض کردیم
$n$
به صورت توانی از 2 است. اگر در هر مرحله‌ی تقسیم ماتریس 
ابعاد ماتریس
توانی از 2 نباشد با افزودن سطرها و ستون‌هایی که همه‌ی درایه‌های آن‌ها صفر است به ماتریس‌ها، آن‌ها را بصورت توانی از 2 در می‌آوریم.
<hr>
</div>

```python
int** addAndSubMatrix(int** A, int** B, int flag, int size){
    // if flag ==1 then add else if flag == -1 sub
    int** C = new int*[size];
    for (int i = 0; i < size; i++) {
        C[i] = new int[size];
    }

    for (int i = 0; i < size; i++) {
        for (int j = 0; j < size; j++) {
            C[i][j] = A[i][j] + (flag) * B[i][j];


        }

    }

    return C;
}
```

```python
int** Strassen(int** A, int** B, int size)
{
    // Suppose variable size is powers of 2 if it's not you fix this problem

    if(size == 2){
        return regularMultiple(A, B, size);
    }
    int newSize = size/2;


    int ** A11 = new int*[newSize];
    for (int i = 0; i < newSize; i++) {
        A11[i] = new int[newSize];
    }

    int ** A12 = new int*[newSize];
    for (int i = 0; i < newSize; i++) {
        A12[i] = new int[newSize];
    }


    int ** A21 = new int*[newSize];
    for (int i = 0; i < newSize; i++) {
        A21[i] = new int[newSize];
    }

    int ** A22 = new int*[newSize];
    for (int i = 0; i < newSize; i++) {
        A22[i] = new int[newSize];
    }


    int ** B11 = new int*[newSize];
    for (int i = 0; i < newSize; i++) {
        B11[i] = new int[newSize];
    }


    int ** B12 = new int*[newSize];
    for (int i = 0; i < newSize; i++) {
        B12[i] = new int[newSize];
    }


    int ** B21 = new int*[newSize];
    for (int i = 0; i < newSize; i++) {
        B21[i] = new int[newSize];
    }

    int ** B22 = new int*[newSize];
    for (int i = 0; i < newSize; i++) {
        B22[i] = new int[newSize];
    }

    for (int i = 0; i < newSize; i++) {
        for (int j = 0; j < newSize; j++) {
            A11[i][j] = A[i][j];
            A12[i][j] = A[i][j + newSize];
            A21[i][j] = A[i + newSize][j];
            A22[i][j] = A[i + newSize][j + newSize];

            B11[i][j] = B[i][j];
            B12[i][j] = B[i][j + newSize];
            B21[i][j] = B[i + newSize][j];
            B22[i][j] = B[i + newSize][j + newSize];
        }
    }


    int** M1 = Strassen(addAndSubMatrix(A11, A22, 1, newSize), addAndSubMatrix(B11, B22, 1, newSize), newSize);
    int** M2 = Strassen(addAndSubMatrix(A21, A22, 1, newSize), B11, newSize);
    int** M3 = Strassen(A11, addAndSubMatrix(B12, B22, -1, newSize), newSize);
    int** M4 = Strassen(A22, addAndSubMatrix(B21, B11, -1, newSize), newSize);
    int** M5 = Strassen(addAndSubMatrix(A11, A12, 1, newSize), B22, newSize);
    int** M6 = Strassen(addAndSubMatrix(A21, A11, -1, newSize), addAndSubMatrix(B11, B12, 1, newSize), newSize);
    int** M7 = Strassen(addAndSubMatrix(A12, A22, -1, newSize), addAndSubMatrix(B21, B22, 1, newSize), newSize);



    int** C11 ;
    C11 = addAndSubMatrix(M1, M4, 1, newSize);
    C11 = addAndSubMatrix(C11, M7, 1, newSize);
    C11 = addAndSubMatrix(C11, M5, -1, newSize);

    int** C12 = addAndSubMatrix(M3, M5, 1, newSize);

    int** C21 = addAndSubMatrix(M2, M4, 1, newSize);

    int ** C22 = addAndSubMatrix(M1, M3, 1, newSize);
    C22 = addAndSubMatrix(C22, M6, 1, newSize);
    C22 = addAndSubMatrix(C22, M2, -1, newSize);

    int ** C = new int *[size];
    for (int i = 0; i < size; i++) {
        C[i] = new int[size];
    }

    for (int i = 0; i < newSize; i++) {
        for (int j = 0; j < newSize; j++) {
            C[i][j] = C11[i][j];
            C[i][j + newSize] = C12[i][j];
            C[i + newSize][j] = C21[i][j];
            C[i+newSize][j+newSize] = C22[i][j];
        }
    }


    return C;

}
```

    $p(x)$
    و

    دو چندجمله‌ای با درجه
    $n$

    $r(x) = p(x).q(x)$
    را محاسبه کنیم.
    <br>

    </h3>

    $$p(x) = p_n x ^ n + p_{n-1} x ^ {n-1} + ... + p_1 x + p_0 $$

    $$q(x) = q_n x ^ n + q_{n-1} x ^ {n-1} + ... + q_1 x + q_0 $$
    در این صورت ضرب این دو به با
    $n^2$
    عمل ضرب و حدودا
    $n^2$
    عمل جمع قابل انجام است.
    <br>
     <strong style="color: #ff6600;">
        تمرین:   
        
    </strong>
    ضرب دو چند جمله‌ای را در قسمت زیر پیاده‌سازی کنید.
    
    
</div>

        تلاش دوم    
    </h3>
    نمایش یک چندجمله‌ای با ضرایبش را می‌شناسیم، طوری که چندجمله‌ای 
    $$p(x) = p_n x ^ n + p_{n-1} x ^ {n-1} + ... + p_1 x + p_0 $$
    با ضرایبش به صورت یکتا مشخص می‌شود. اما راه دیگری هم برای نمایش یک چندجمله‌ای وجود دارد. به عنوان مثال، یک خط با هر دو نقطه روی آن
    به صورت یکتا تعیین می‌شود.
    یک چندجمله‌ای درجه دو با سه نقطه روی آن به صورت یکتا مشخص می‌شود. به طور کلی هم قضیه زیر برقرار است:
    <br>
    <strong style="color: #ff6600;">
        قضیه:   
    </strong>
    یک چندجمله‌ای درجه
    $n$
    (با ضرایب مختلط) با
    $n+1$
    نقطه روی آن به صورت یکتا مشخص می‌شود.
    <br>
    برقرار بودن قضیه بالا برای اعداد مختلط نه تنها از نظر ریاضی درست است، بلکه در ادامه برای حل مسئله مورد نظرمان لازم هم خواهد بود.  
    <br>
    با در دست داشتن قضیه بالا، می‌توان نمایش دیگری برای یک چند جمله‌ای پیدا کرد به این صورت که به جای آنکه چندجمله‌ای را با ضرایبش مشخص کنیم، با تعدادی نقطه روی آن
    (یعنی تعدادی
    $p(x_i)$
    برای
    $x_i$
    های مختلف)
    مشخص می‌کنیم.
    انجام ضرب با این نمایش چند جمله‌ای بسیار ساده است، چون که با داشتن 
    $p(x_i)$
    و
    $q(x_i)$
    می‌توان مقدار
    $p(x_i)q(x_i)$
    را که متعلق به حاصلضرب دو چندحمله‌ای است را محاسبه کرد.
    <br>
    <strong  style="color: #ff6600;">نکته:</strong>
    چندجمله‌ای حاصلضرب یک چندجمله‌ای از درجه
    $2n$ 
    است و باید حاصلضرب 
    $2n$
    نقطه را به دست آورد.
   <br>
    خب طبق آنچه در بالا گفته شد، با داشتن تعداد کافی نقطه از دو چندجمله‌ای، محاسبه حاصلضرب‌شان با 
    $2n$
    ضرب قابل انجام است.
    اما ما معمولا نمایش عادی یک چند جمله‌ای را داریم، پس سوال این است که هزینه تبدیل چندجمله‌ای به این نمایش چقدر است؟
    شکل زیر خلاصه گفته‌ها تا به اینجا را نشان می‌دهد:
    <img src='diagram.png'>
    شکل نشان می‌دهد که روش جدیدی برای محاسبه حاصلضرب پیدا کرده‌ایم اما اگر تبدیل را به صورت بهینه‌ای انجام ندهیم، بهبودی حاصل نشده و پیچیدگی ضرب همچنان 
    $O(n^2)$
    است.
    اگر بتوانیم راهی سریع‌تر برای درونیابی و برونیابی چندجمله‌ای پیدا کنیم می‌توانیم ضرب را سریع‌تر از حالت معمول انجام دهیم.
    این تبدیل نمایش ضرایب به نقاط همان 
    <strong> تبدیل فوریه </strong>
    نام دارد.
</div>    

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <h2 style="color: #ff6600;">
تبدیل فوریه    
</h2>
    طبق آنچه در بالا گفتیم، صورت مسئله ساده شده به این ترتیب است:
    با داشتن نمایش یک چندجمله‌ای درجه 
    $n$
    با ضرایب به شکل
        $$p(x) = p_{n-1} x ^ {n-1} + p_{n-2} x ^ {n-1} + ... + p_1 x + p_0 $$
    مقدار این چندجمله‌ای را در 
    $n$
    نقطه به دست آوردید.
    <br>
    شاید به نظر بیاید جایی برای بهبود سرعت این محاسبات نداشته باشیم، اما مقادیر
    $x_i$
    را می‌توانیم تعیین کنیم. پس سعی می‌کنیم مقادیری انتخاب کنیم که بتوان محاسبات را سریع‌تر انجام داد.
    از آنجایی که نمی‌توانیم 
    $x_i$
    یکسان انتخاب کنیم، ساده‌ترین جایگزین 
    $x_i = - x_j$
    به نظر می‌رسد که شاید بخشی از محاسبات را کم کند.
      ( فرض کنید بدون کاهش کلیت مسئله $n$ فرد باشد.) چند جمله‌ای را به صورت زیر نشان می‌دهیم:
    $$ p(x) = p_{n} x ^ {n} + p_{n-1} x ^ {n-1} + p_{n-2} x ^ {n-2} + ... + p_1 x + p_0 $$
    $$   = (p_{n} x ^ {2k + 1} + p_{n-2} x ^ {2k-1} + p_{n-4} x ^ {2k - 3} + ... + p_1 x) $$
    $$   + (p_{n-1} x ^ {2k} + p_{n-3} x ^ {2k - 2} + ... + p_0) $$
    $$   = x \times P_0(x^2) +  P_1(x^2)$$
    که در فرمول بالا
    $$P_0(x) = p_{n} x ^ {k} + p_{n-1} x ^ {k-1}... + p_1 $$
    $$p_1(x) = p_{n-1} x ^ {k} + ... + p_0 $$
    <strong style="color: #ff6600;">
        تمرین:   
    </strong>
    درستی روابط بالا را اثبات کنید.
    <br>
    آنچه از روابط بالا باید دریافت این است که برای 
    $x_i$
    و 
    $x_j = - x_i$
    داریم:
    
    $$P_0(x_{i}^2) = P_0(x_{j}^2)$$ 
    $$P_1(x_{i}^2) = P_1(x_{j}^2)$$
   
    
    <br>
    اگر بخواهیم دقیق‌تر بگوییم، فرض کنید 
     $ x_0, x_1, ... , x_{n-1} $ طوری باشند که نیمه اول قرینه نیمه دوم باشد
    ($x_{i+n/2} = -x_i$)
    پس محاسبه 
    $p(x)$
    در این
    $n$
    نقطه برابر است با محاسبه 
    $p_0(x)$
    و
    $p_1(x)$
    در
    $x_0^2, x_1^2, ... , x_{n_1}^2$
    به انضمام چند عملیات اضافه که تعدادشان از 
    $O(n)$
    است.
    اما نیمه اول قرینه نیمه دوم است! پس کافیست 
    $p_0(x)$
    و
    $p_1(x)$
    را فقط در نصف نقاط حساب کنیم! و اینگونه به دو زیر مسئله با اندازه کوچک‌تر دست پیدا می‌کنیم.
    (چون
    $p_0(x)$
    و
    $p_1(x)$
    از درجه
    $\frac{n}{2}$
    هستند و میخواهیم در
    $\frac{n}{2}$
    نقطه مقدارشان را پیدا کنیم.
    <br>
    به نظر می‌آیند مسئله به طور کامل به دو زیرمسئله شکسته شده، تنها چیزی که مانده، این است که مقادیر
    $x_i$
    ها را تعیین کنیم.
    در مرحله اول 
    $n$
    عدد داشتیم که نیمه اول مثبت بود و نیمه دوم منفی. اما در مرحله دوم چه؟ مگر همه اعداد به شکل
    $x_i^2$
    نیستند؟؟  پس چطور می‌توان نصف را مثبت و نصف دیگر را منفی انتخاب کرد؟؟
    <br>
    جواب این سوال را بالاتر به صورت ضمنی داده‌ایم. استفاده از اعداد مختلط!
    اعدادی که شرایط مورد نظر برای حل مسئله را دارند، ریشه‌های
    $n$
    ام واحد هستند.
    <br>
    به صورت خلاصه ریشه‌های 
    $n$
    ام واحد اعدادی هستند که در معادله
    $$
    \omega ^ n - 1 = 0
    $$
    صدق می‌کنند. توضیحات درباره خواص اعداد مختلط و اینکه چرا این اعداد ویژگی‌های مورد نظر ما را دارند این دفترچه را طولانی خواهد کرد.
    برای همین فقط الگوریتم بالا را برای
    $x_i$
    های مناسب توضیح می‌دهیم.
    <br>
    اگر
    $\omega$
    برابر ریشه 
    $n$
    ام واحد باشد،
    آنوقت
    $x_i = \omega ^ i$
    قرار می‌دهیم.
    آنوقت می‌خواهیم مقدار 
    $p(x)$
    را در 
    $\omega^0, \omega^1,..., \omega^{n-1}$
    حساب کنیم. برای اینکار با استفاده از معادله
    $$
    p(x) = p_0(x^2) + x p_1(x^2)
    $$
    ابتدا مقدار
    $p_0(x)$
    و
    $p_1(x)$
    را در 
    $(\omega^0)^2, (\omega^1)^2,..., (\omega^{n-1})^2$
    حساب می‌کنیم. اما با کمی محاسبه می‌توان دید که این‌ها درواقع فقط
    $n/2$
    عدد متمایز یا همان ریشه‌های 
    $n/2$
    واحد هستند.
    <br>
    <strong style="color: #ff6600;">
        نکته:   
    </strong>
    ممکن است فکر کنید که پیاده‌سازی اعداد مختلط چگونه خواهد بود. اما هر عدد مختلط به صورت یک زوج مرتب قابل نمایش است و ضرب و جمع اعداد مختلط هم به راحتی با همان نمایش با زوج مرتب قابل انجام است. در زبان
    C
    این نوع داده پیاده‌سازی شده است و به صورت زیر قابل استفاده است.
    
</div>    

```python
typedef complex<double> cd; 
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    با کمک این نوع داده می‌توان تبدیل فوریه را پیاده‌سازی کرد.
</div>    

```python
// Recursive function of FFT 
std::vector<cd> fft(std::vector<cd>& p) 
{ 
    int n = p.size(); 
  
    // Recursion base case: if input contains just one element 
    if (n == 1) 
        return vector<cd>(1, p[0]); 
  
    // Calculating roots of unity
    vector<cd> w(n); 
    for (int i = 0; i < n; i++) { 
        double alpha = 2 * M_PI * i / n; 
        w[i] = cd(cos(alpha), sin(alpha)); 
    } 
  
    vector<cd> p0(n / 2), p1(n / 2); 
    for (int i = 0; i < n / 2; i++) { 
        // first subproblem: even indexed coefficients 
        p0[i] = p[i * 2]; 
        // second subproblem: odd indexed coefficients 
        p1[i] = p[i * 2 + 1]; 
    } 
  
    // Recursive call for first subproblem 
    vector<cd> y0 = fft(p0);  
    // Recursive call for second subproblem 
    vector<cd> y1 = fft(p1); 
  
    // for storing values of y0, y1, y2, ..., yn-1. 
    std::vector<cd> y(n); 
  
    for (int k = 0; k < n / 2; k++) { 
        y[k] = y0[k] + w[k] * y1[k]; 
        y[k + n / 2] = y0[k] - w[k] * y1[k]; 
    } 
    return y; 
} 
```

```python
{ 
    int n=3;
    vector<cd> poly_p; // p(x) = 3x^2 + 2x + 1
    poly_p.push_back(1);
    poly_p.push_back(2);
    poly_p.push_back(3);
    
    vector<cd> fft_p = fft(poly_p);
    for (int i=0;i<n;i++){
        cout << fft_p[i];
    }
}
```

    <h2 style="color: #ff6600;"> تحلیل زمانی </h2>
    ابتدا پیچیدگی زمانی تبدیل فوریه را به شرح زیر می‌توانیم حساب کنیم:
    $$
        T(n) = 2 T(n/2) + O(n)
    $$
    که طبق قضیه اصلی می‌دانیم که جواب برابر خواهد بود با:
    $$
        T(n) = O(n\log n)
    $$
    به عنوان تمرین هم می‌توانید پیچیدگی زمانی ضرب دو چند جمله‌ای را که متشکل از یک تبدیل فوریه و یک تبدیل عکس فوریه است، حساب کنید.
