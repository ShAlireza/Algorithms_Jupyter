<div style="font-size: 22px; color: #ff6600; font-weight: bold">
مقدمه
</div><br>
در بخش‌های قبلی درس بررسی کردیم که اگر بخواهیم چند ماتریس را در‌هم ضرب کنیم چگونه بهترین توالی برای این ضرب‌ها را بدست بیاوریم تا سریعتر به جواب برسیم و به روش ضرب دو ماتریس کاری نداشتیم.
<br>
در این بخش چگونگی ضرب دو ماتریس را مورد بررسی قرار می‌دهیم و به دنبال راه‌حلی سریع و بهینه برای این کار هستیم.
</div>
<hr>

<div id="section2" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
بخش اول: الگوریتم استراسن برای ضرب دو ماتریس
</div><br>            
 دو ماتریس 
$A_{n*n}$
و
$B_{n*n}$
داریم می‌خواهیم 
$C_{n*n} = AB$
را حساب کنیم.
<br>

<hr>
</div>    

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
راه حل اول(راه حل بدیهی)
</div>         
طبق تعریف ضرب ماتریس‌ها عمل کرده و هر سطر از ماتریس 
$A$
را در هر ستون از ماتریس 
$B$
ضرب می‌کنیم.
<img src="matrix_multiplication.png" style="width:60%">
<strong>
    پیچیدگی زمانی:
</strong>
چون برای محاسبه هر درایه از ماتریس
$C$
باید یک ضرب سطر در ستون انجام شود که هر کدام از مرتبه زمانی 
$O(n)$
زمان می‌برد.
و چون ماتریس 
$C$
دارای
$n^2$
درایه است, انجام ضرب با این روش در مرتبه‌ی زمانی 
$O(n^3)$
انجام می‌شود.
<hr>
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
راه حل دوم (الگوریتم استراسن)
</div>            
ایده اصلی این الگوریتم تقسیم و حل است. دو ماتریس 
$n*n$
$A$
و
$B$
داده‌شده‌است. با فرض بر اینکه عدد 
$n$
به صورت توانی از عدد
$2$
است.
$(n = 2 ^ k and ‌  k \in N )$
هر کدام از دو ماتریس
$A$
و
$B$
را به 4 زیرماتریس تقسیم می‌کنیم که هر کرام 
$\frac{n}{2} * \frac{n}{2} $
هستند.
(هر کدام از 
$A_{ij}$
ها و
$B_{ij}$
ها ماتریس هستند.)
<br/>


\begin{equation}
A=
\begin{pmatrix}
A_{11} & A_{12} \\
A_{21} & A_{22}
\end{pmatrix}
,
B=
\begin{pmatrix}
B_{11} & B_{12} \\
B_{21} & B_{22}
\end{pmatrix}
\end{equation}


<hr>
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
</div>            
به راحتی می‌توان اثبات کرد:
\begin{equation}
C=
\begin{pmatrix}
C_{11} & C_{12} \\
C_{21} & C_{22}
\end{pmatrix}
=
\begin{pmatrix}
A_{11}B_{11} + A_{12}B_{21} & A_{11}B_{12} + A_{12}B_{22} \\
A_{21}B_{11} + A_{22}B_{21} & A_{21}B_{12} + A_{22}B_{22}
\end{pmatrix}
\end{equation}
<img src="strassen_new.png" style="width:50%">

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
در اینجا در هر مرحله باید 8 ضرب ماتریس‌های 
$\frac{n}{2} * \frac{n}{2} $
را حساب کنیم, یعنی:
\begin{equation}
T(n) = 8T(\frac{n}{2}) + O(1)
\end{equation}
که با محاسبه با قضیه اصلی بدست می‌آید:
$T(n) \in O(n^3)$
<br/>
می‌بینیم که زمان اجرای ضرب هنوز بهبود پیدا نکرد. در این روش به‌جای محاسبه این 8 عبارت  
7 عبارت جدید را محاسبه می‌کنیم.
که برای محاسبه هر کدام از این 7 عبارت به محاسبه یک ضرب ماتریس 
$\frac{n}{2} * \frac{n}{2} $
نیاز است.
<br/>
این 7 عبارت به شرح زیر است:
\begin{equation}
M_1 = (A_{11} + A_{22})(B_{11} + B_{22})
\end{equation}
\begin{equation}
M_2 = (A_{21} + A_{22})(B_{11})
\end{equation}
\begin{equation}
M_3 = (A_{11})(B_{12} - B_{22})
\end{equation}
\begin{equation}
M_4 = (A_{22})(B_{21} - B_{11})
\end{equation}
\begin{equation}
M_5 = (A_{11} + A_{12})(B_{22})
\end{equation}
\begin{equation}
M_6 = (A_{21} - A_{11})(B_{11} + B_{12})
\end{equation}
\begin{equation}
M_7 = (A_{12} - A_{22})(B_{21} + B_{22})
\end{equation}
و زیر‌ماتریس‌های 
$C$
به صورت زیر محاسبه می‌شوند:
\begin{equation}
C_{11} = M_1 + M_4 + M_7 - M_5 \\
C_{12} = M_3 + M_5 \\
C_{21} = M_2 + M_4 \\ 
C_{22} = M_1 + M_3 + M_6 - M_1
\end{equation}

<strong>
    نکته:
</strong>
در ابتدای این روش فرض کردیم
$n$
به صورت توانی از 2 است. اگر هم 
$n$
توانی از 2 نباشد با افزودن سطرها و ستون‌های صفر به ماتریس‌ها آنها را بصورت توانی از 2 در می‌آوریم.
<hr>
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
تحلیل زمانی این راه حل
</div>            
 \begin{equation}
T(n) = 7T(\frac{n}{2}) + O(1)
\end{equation}
<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
که طبق قضیه اصلی:
$T(n) \in O(n^{\lg{7}}) \approx O(n^{2.81})$
</br>
<br>
می‌بینیم که زمان بهتر شد.
</br>
<br>

    پرسش:
</strong>
