<h3>مقدمه</h3>

    در بخش قبلی سعی می‌کردیم الگوریتم‌ها با ضرایب تقریب متفاوت ارائه کنیم. اما بعضاً ممکن است بخواهیم دقت الگوریتم را خودمان تعیین کنیم. برای مثال فرض کنید سیستمی قوی داریم و می‌توانیم هزینه بیشتری از نظر زمانی بدهیم و در عوض دقت بیشتری بگیریم. و یا بالعکس.
    بنابراین می‌خواهیم پارامتری تحت عنوان دقت را به صورت یک **متغیر** ارائه دهیم.
    در این بخش با چنین الگوریتم‌هایی سر و کار داریم.
    <br>
    تعریف: طرح‌های تقریبی چند‌جمله‌ای خانواده‌ای از الگوریتم‌ها مثل $A_{\epsilon}$
    است.
    به طوریکه به‌ازای هر ضریب 
    $\epsilon > 0$
    الگوریتم چندجمله‌ای 
    $(1+\epsilon)-approx$
    است (برای مسائل کمینه‌سازی)
    و 
    $(1-\epsilon)-approx$
    است (برای مسائل بیشینه‌سازی)
    <br>
    توجه کنید ممکن است زمان الگوریتم ما وابسته به این $\epsilon$ باشد.
    برای مثال ممکن است الگوریتمی با 
    $O(n^{\frac{1}{\epsilon}})$
    داشته‌باشیم. در اینصورت اگر مقدار 
    $\epsilon$
    خیلی کم باشد (یعنی الگوریتمی با دقت زیاد داشته‌باشیم) آن وقت زمان الگوریتم بیشتر شود.
    این الگوریتم بر حسب $n$ چند جمله‌ای است ولی برحسب
    $\frac{1}{\epsilon}$
    چندجمله‌ای نیست!
    <br>
    تعریف: طرح‌های تقریبی کاملا چندجمله‌ای یا FPTAS دسته‌ای از الگوریتم‌های PTAS هستند که هم بر اساس اندازه ورودی و هم بر اساس پارامتر دقت یعنی
    $\epsilon$
    چند‌جمله‌ای است.
    <br>
    معمولاً پیدا کردن FPTAS در مسائل بسیار خوشحال کننده‌است.

<h3>طرح های تقریبی چندجملەای</h3>

</font></div>

<div style="direction:rtl;line-height:300%;" id="section2"><font face="XB Zar" size=5>
<h3>کولەپشتی</h3>
به عنوان ورودی n شی داریم که هر کدام ارزش $p_{i}$ و اندازه ی $s_{i}$ دارند. هدف، قرار دادن این اشیا در کوله‌پشتی با ظرفیت B به صورتی است که مقدار ارزش بیشینه حاصل شود. (برای سادگی فرض می‌کنیم برای تمامی اشیاء $s_{i} < B$ و باید زیرمجموعه ای از این اشیاء پیدا کنیم $I \subset [n]$ تا مقدار $\sum_{i\in I}^{}p_{i}$ را ماکسیمم کند و شرط مسئله $\sum_{i\in I}^{}s_i\leqslant B$  است.)
<br/>
از آنجا که مسئله ی کوله پشتی از مسئله ی تقسیم بندی ( partition problem  ) کاهش می‌یابد بنابراین جز np_hard است.
<br/>
استفاده از الگوریتم پویا برای حل این مسئله به ما جواب دقیق را در زمان O(poly(n,pmax) $\times$ log(nB)) به ما می دهد. اما این مسئله از نظر اندازه چندجمله ای نیست.
<br/>
بنابراین ما از الگوریتم های تقریبی برای حل این مسئله استفاده می‌کنیم.
<br/>
<font face="XB Zar" size=6>
استفاده از الگوریتم حریصانه 1:
<br/>    
<font face="XB Zar" size=5>    
1 : اشیا را به صورت نزولی برحسب $\frac{p_i}{s_i}$ مرتب کن.
<br/>
2 : به صورت حریصانه اشیاء را به ترتیب بر دار
<br/>
این الگوریتم، الگوریتم خوبی نیست و در بسیاری از مثال ها جواب بدی به ما می دهد. مانند مثال روبرو ( شی یک : s = 1, p = 2 و شی دو : s = B, p = B )
<br/>
<font face="XB Zar" size=6>    
استفاده از الگوریتم حریصانه 2 (redux) :
<font face="XB Zar" size=5> 
<br/>
1 : اشیا را به صورت نزولی برحسب $\frac{p_i}{s_i}$ مرتب کن.
<br/>
2 : به صورت حریصانه شیء ها را اضافه کن تا زمانی که به شی‌ای با $a_i$ بسیار بزرگ برسیم. $(\sum_{k=1}^{i} s_i > B)$
<br/>
3 : از بین $\left \{ a_1,a_2,...,a_{i-1} \right \}$ و $\left \{ a_i \right \}$ هر کدام بهتر بود را انتخاب می‌کنیم.
<br/>
قضیه : الگوریتم حریصانه ی redux دو_تقریب برای مسئله ی کوله پشتی است.
<br/>
اثبات : ما برای این مسئله، یک  الگوریتم حریصانه پیشنهاد دادیم. بنابراین می توانیم بگوییم، اگر الگوریتم ما زیربهینه (suboptimal) باشد در آخر یک فضای B-S باقی می ماند. اگر الگوریتم ما می‌توانست قسمتی از اشیاء را درون کوله پشتی قرار دهد، آنگاه با اضافه کردن $\frac{B-S}{s_{k-1}}p_{k+1}$  به کوله پشتی ما یا به OPT می‌رسیدیم، یا به کوله پشتی پرشده. ( دقت کنید در این مسئله نمی‌توانیم از بخشی از اشیاء استفاده کنیم.)
<br/>
بنابراین یا $\sum_{k=1}^{i-1}p_i \geqslant \frac{1}{2}OPT$ یا $p_{k+1} \geqslant \frac{B-S}{s_{k+}} p_{k+1} \geqslant \frac{1}{2}OPT$ اتفاق می‌افتد.
<br/>
<font face="XB Zar" size=6>    
تقریب $1+\epsilon $ :
<font face="XB Zar" size=5>   
<br/>
اگر به الگوریتم پویای ارائه شده دقت کنیم می‌بینیم مشکل این الگوریتم این بود که در زمان 
O(poly(n, pmax))  انجام می‌شد. بنابراین اگر ما مقیاس ها را کم کنیم، آنگاه مشکل ما حل می‌شود. الگوریتم fptas for knapsack(fully polynomial time approximation scheme) به این ترتیی عمل می کند که توسط kim و Ibarra ارائه شد.
<br/>
///
<br/>
مجموعه ی s’ که توسط الگوریتم بالا به دست می آید در شرط $profit(s') \geqslant (1-\epsilon )\times profit(s^{*})$ صدق می کند که $s^{*}$ در آن جواب بهینه است. بنابر این این الگوریتم 1_تقریب است.
