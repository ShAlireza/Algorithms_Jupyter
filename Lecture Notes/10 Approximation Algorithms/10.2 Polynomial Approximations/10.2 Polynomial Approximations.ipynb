مقدمه
</div>
    در بخش قبلی دیدید یک مسئله ممکن است تقریب‌های متفاوتی داشته‌باشد برای مثال برای فروشنده دوره‌گرد ابتدا یک ۲-تقریب و سپس یک ۱.۵ تقریب ارائه کردیم.
    سؤال منطقی‌ای که پیش می‌آید این است که این روند تا کجا ادامه پبدا می‌کند آیا ممکن است الگوریتمی با تقریب بهتر ارائه کنیم.
    <br/>
    گاهی ممکن است تن به محاسبات بیشتری بدهیم تا دقت الگوریتم را بیشتر کنیم برای مثال ممکن است یک ۱۰تقریب داشته باشیم که با هزینه زمانی کمتری حل شود اما از طرفی حاضر
    باشیم هزینه زمانی بیشتری بدهیم که این تقریب ۲تقریب بشود.
    <br/>
    در این بخش ابتدا تعریفی را ارائه می‌دهیم و سپس یک مسئله معروف 
    «کوله‌پشتی»
    با چند روش برای حل کردن آن به‌دست می‌آوریم.
    سپس بک طرح تقریبی چند‌جمله‌ای 
    ارائه می‌کنیم که بتوان با هر دقتی مسئله را حل کرد.

<div style="font-size: 22px; color: #ff6600; font-weight: bold">
طرح‌های تقریبی چند‌جمله‌ای
</div>
    در بخش قبلی سعی می‌کردیم الگوریتم‌ها با ضرایب تقریب متفاوت ارائه کنیم. اما بعضاً ممکن است بخواهیم دقت الگوریتم را خودمان تعیین کنیم. برای مثال فرض کنید سیستمی قوی داریم و می‌توانیم هزینه بیشتری از نظر زمانی بدهیم و در عوض دقت بیشتری بگیریم. و یا بالعکس.
    بنابراین می‌خواهیم پارامتری تحت عنوان دقت را به صورت یک **متغیر** ارائه دهیم.
    در این بخش با چنین الگوریتم‌هایی سر و کار داریم.
    <br>
    تعریف:
    (PTAS)
    طرح‌های تقریبی چند‌جمله‌ای خانواده‌ای از الگوریتم‌ها مثل $A_{\epsilon}$
    است.
    به طوریکه به‌ازای هر ضریب 
    $\epsilon > 0$
    الگوریتم چندجمله‌ای 
    $(1+\epsilon)-approx$
    است (برای مسائل کمینه‌سازی)
    و 
    $(1-\epsilon)-approx$
    است (برای مسائل بیشینه‌سازی)
    <br>
    توجه کنید ممکن است زمان الگوریتم ما وابسته به این $\epsilon$ باشد.
    برای مثال ممکن است الگوریتمی با 
    $O(n^{\frac{1}{\epsilon}})$
    داشته‌باشیم. در اینصورت اگر مقدار 
    $\epsilon$
    خیلی کم باشد (یعنی الگوریتمی با دقت زیاد داشته‌باشیم) آن وقت زمان الگوریتم بیشتر شود.
    این الگوریتم بر حسب $n$ چند جمله‌ای است ولی برحسب
    $\frac{1}{\epsilon}$
    چندجمله‌ای نیست!
    <br>
    تعریف:
    (FPTAS)
    طرح‌های تقریبی کاملا چندجمله‌ای یا FPTAS دسته‌ای از الگوریتم‌های PTAS هستند که هم بر اساس اندازه ورودی و هم بر اساس پارامتر دقت یعنی
    $\epsilon$
    چند‌جمله‌ای است.
    <br>
    معمولاً پیدا کردن FPTAS در مسائل بسیار خوشحال کننده‌است.
    زیرا می‌توانیم با توجه با سیستمی که در اختیار داریم یک
    TradeOff
    بین دقت و 
    performance
    داشته‌باشیم.

<div style="font-size: 22px; color: #ff6600; font-weight: bold">
مسئله کوله‌پشتی
</div>
    <div id="section2.1">
    <div style="font-size: 17px; color: #ff6600; font-weight: bold">
        معرفی مسئله
    </div>
    مسئله کوله‌پشتی مسئله معروفی است که تا الآن با آن آشنایی دارید اما صرفاُ جهت یاد‌آوری صورت مسئله را در زیر بیان می‌کنیم:
    <br>
به عنوان ورودی n شی داریم که هر کدام ارزش $s_{i}$ و اندازه ی $p_{i}$ دارند. هدف، قرار دادن این اشیا در کوله‌پشتی با ظرفیت B به صورتی است که مقدار ارزش بیشینه حاصل شود. (برای سادگی فرض می‌کنیم برای تمامی اشیاء $p_{i} < B$ و باید زیرمجموعه ای از این اشیاء پیدا کنیم $I \subset [n]$ تا مقدار $\sum_{i\in I}^{}s_{i}$ را ماکسیمم کند و شرط مسئله $\sum_{i\in I}^{}p_i\leqslant B$  است.)
<br/>
    سؤال منطقی‌ای که اول بوجود می‌آید این‌است که: 
    «آیا واقعا لازم است سراغ راه‌حل های تقریبی برویم؟ آیا راه‌چند جمله‌ای برای حل این مسئله هست‌یا نه؟»
    <br>
از آنجا که مسئله ی کوله پشتی از مسئله ی تقسیم بندی ( partition problem  ) کاهش می‌یابد بنابراین جز NP_Hard است.
<br/>


<div id="section2.2" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 17px; color: #ff6600; font-weight: bold">
        روش‌برنامه‌ریزی پویا
    </div>
    در ابتدا راه‌حلی ارائه می‌دهیم که با استفاده از برنامه‌نویسی پویا انجام پذیر است. فرض کنید 
    به ازای هر 
    $i, j$
    که
    $0 \le i \le n$
    و
    $0 \le j \le p_{max}$
    که در آن
    $p_{max}$
    اندازه بیشترین مقدار
    ممکن برای اندازه جواب است یعنی
    $p_{max} = max_{I \subset [n]} : (\sum_{i \in I} p_i) \le B$
    فرض کنید
    یک
    «دانای کل»
    داریم به فرم
    $O(i, j)$
    که می‌گوید اگر ظرفیت کوله‌پشتی ما 
    $j$
    باشد و بخواهیم صرفا از اشیای ۱ تا
    $i$
    استفاده کنیم ماکسیمم ارزشی که می‌توانیم برداریم چیست.
    <br/>
    حالا می‌توانیم به‌شکل زیر 
    $O(i, j)$
    را از روی یکسری
    $O(i', j')$
    های دیگر به‌دست بیاوریم:
    <br/>
    <li>
    اگر 
    $p_i \le j$
    باشد آنگاه:
    $O(i,j) = max(O(i-1, j), O(i-1, j - p_i) + s_i)$
    یعنی می‌توانیم انتخاب کنیم که شئ $i$
    ام را برداریم یا برنداریم.
    </li>
    <li>
    در غیر اینصورت:
    $O(i, j) = O(i-1, j)$
    زیرا نمی‌توانیم شئ $i$ ام را برداریم.
    </li>
    <br/>
    برای پایه هم می‌توانیم فرض کنیم
    :
    $\forall_{k \in [0, p_{max}]} : O(0, k) = 0$
    کد زیر چیزی که ما گفتیم را پیاده‌سازی می‌کند:


```python
#include <bits/stdc++.h>
using namespace std;

const int MAX_B = 2020, MAX_N = 2020;
int dp[MAX_N][MAX_B];

struct Obj{
	int s, p;
}obj[MAX_N];

// solve objects in [l, r) with capacity B
long long solve_using_dp(int n, int B) {
	fill(dp[0], dp[0] + B + 1, 0);
	for (int i = 0; i < n; i++)
		for (int j = B; j >= 0; j--) {
			if (j >= obj[i].p)
				dp[i + 1][j] = max(dp[i][j], dp[i][j - obj[i].p] + obj[i].s);
			else
				dp[i + 1][j] = dp[i][j];
		}
	return dp[n][B];
}

int main() {
	int n, B;
	cin >> B >> n;
	for (int i = 0; i < n; i++) 
		cin >> obj[i].p >> obj[i].s;
	// exclude all with size larger than B
	for (int i = 0; i < n; i++) {
		if (obj[i].p > B) {
			swap(obj[i], obj[n-1]);
			n--;
		}
	}
	cout << solve_using_dp(n, B) << endl;
	return 0;
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
می‌توانیم به‌راحتی ببینیم که این الگوریتم
$O(np_{max})$
است. ممکن است گول ظاهر این تحلیل را بخورید و نتیجه بگیرید که مسئله 
NP_Hard
نیست!
(یا شاید‌هم باید جایزه میلیون دلاری 
P = NP
را دریافت کنید!)
    در واقع ما الگوریتمی از
    $O(poly(n, p_{max}))$
    به دست آوردیم
اما واقعیت این‌است که تحلیل زمانی الگوریتم‌ها بر اساس اندازه‌ورودی در نظر گرفته می‌شود.
<br/>
به‌عبارت بهتر شما برای ذخیره‌سازی هر 
$p_i$
به اندازه
$log(p_i)$
بیت ورودی می‌گیرید و زمان الگوریتم در واقع
$O(n \times radix ^ {log(p_i)})$
است که براساس اندازه ورودی نمایی است!
<br/>
در واقع به اینگونه مسائل مسائل
Psuedo Polynomial
نیز می‌گویند.

       راه‌حل با الگوریتم حریصانه
    </div>
        توجه کنید اساس طراحی الگوریتم حریصانه پیدا‌کردن پارامتر خوب است
        ممکن است ترتیبمان این‌باشد که اشیا با ارزش‌تر را زود‌تر برداریم اما می‌توان به‌سادگی دید چنین روشی برای الگوریتم حریصانه مفید نیست!
        ممکن است چیزی که برمی‌داریم خیلی وزین باشد و بهتر باشد اشیا کم ارزش ولی کوچک برداریم تا بتوانیم مقدار زیادی از آن‌ها داشته‌باشیم.
        <br/>
        از طرفی توجه کردن تنها به اندازه هم به ما الگوریتم‌های غلطی می‌دهد لذا بهترین کار چیزی مابین این‌دو است یعنی
        چیزی را برداریم که 
        «چگالی»
        ارزشش بیشینه باشد
        و به ازای هر واحد اندازه بیشترین ارزش را بتوانیم برداریم.
        <br/>
    <div style="font-size: 17px; color: #ff6600;">
استفاده از الگوریتم حریصانه 1:
    </div> 
<li>
    اشیا را به صورت نزولی برحسب $\frac{s_i}{p_i}$ مرتب کن.
    </li>
<li>
    به صورت حریصانه اشیاء را به ترتیب بر دار
    </li>
    اما متأسفانه حتی این‌روش هم روش خوبی نیست در واقع مثالی ارائه می‌دهیم که به‌ازای
    $B$ه
    های بزرگ بد عمل می‌کند.
    <br/>
    $$ p = <1, B>$$
    $$s = <2, B>$$
    طبق الگوریتم عنصر اول را برمی‌داریم و دیگر چیزی برنمی‌داریم اما جواب بهینه $B$ 
    است و جواب ما $2$
    است در نتیجه هیچ ضریب اطمینانی نداریم!

استفاده از الگوریتم حریصانه 2 (redux) :
</div>  
<li>
اشیا را به صورت نزولی برحسب $\frac{s_i}{p_i}$ مرتب کن.
</li>
<li>
به صورت حریصانه شیء ها را اضافه کن تا زمانی که دیگر نتوانیم شیئی اضافه کنیم. $(\sum_{k=1}^{i} p_i > B)$
</li>
<li>
در این‌صورت یا همان اشیا را انتخاب می‌کنیم یا آخرین عضوی را که دیگر جا نمی‌شد را به تنهایی انتخاب می‌کنیم.
</li>
    در نگاه اول صرفاً بنظر می‌رسد راهی برای پیشگیری از صرفاً مثال نقض سری پیش است اما در کمال تعجب این الگوریتم واقعاً یک تقریب خوب به ما می‌دهد.
    <br/>
<span style="color: red">قضیه:</span>
    الگوریتم حریصانه ی redux دو_تقریب برای مسئله ی کوله پشتی است.
<br/>
اثبات : ما برای این مسئله، یک  الگوریتم حریصانه پیشنهاد دادیم. بنابراین می توانیم بگوییم، اگر الگوریتم ما زیربهینه (suboptimal) باشد در آخر یک فضای B-P باقی می ماند. اگر الگوریتم ما می‌توانست قسمتی از اشیاء را درون کوله پشتی قرار دهد
    برای مثال یک چهارم شیئی را قرار دهد
    ، آنگاه با اضافه کردن $\frac{B-P}{p_{i}}s_{i}$  به کوله پشتی ما یا به OPT می‌رسیدیم، یا به کوله پشتی پرشده. ( دقت کنید در این مسئله نمی‌توانیم از بخشی از اشیاء استفاده کنیم.)
<br/>
بنابراین یا $\sum_{k=1}^{i-1}s_i \geqslant \frac{1}{2}OPT$ یا $s_{i} \geqslant \frac{B-P}{p_{i}} s_{i} \geqslant \frac{1}{2}OPT$ اتفاق می‌افتد.
<br/>
    <img src="img/Greed2.jpg" width = 400 />
    از دیگر فواید این روش کد کوتاه است!
<div/>                              

```python
#include <bits/stdc++.h>
using namespace std;

const int MAX_N = 2020;

struct Obj{
	int s, p;
}obj[MAX_N];

bool cmp(const Obj &a, const Obj &b) {
	return 1ll * a.s * b.p > 1ll * b.s * a.p;
}

// return the approximation for objects in interval [l, r)
long long redux(int n, int B) { 
	long long return_value = 0;
	long long sum_of_s = 0;
	int remaining_space = B;
	for (int i = 0; i < n; i++) {
		remaining_space -= obj[i].p;
		if (remaining_space < 0) {
			return_value = max((long long)obj[i].s, sum_of_s);
			break;
		}
		sum_of_s += obj[i].s;
	}
	return max(return_value, sum_of_s);
}

int main() {
	int n, B;
	cin >> B >> n;
	for (int i = 0; i < n; i++) 
		cin >> obj[i].p >> obj[i].s;
	// exclude all with size larger than B
	for (int i = 0; i < n; i++) {
		if (obj[i].p > B) {
			swap(obj[i], obj[n-1]);
			n--;
		}
	}
	
	// sort object in decending order of: s[i]/p[i]
	sort(obj, obj + n, cmp);
	
	cout << redux(n, B) << endl;
	return 0;
}
```

    <div style="font-size: 17px; color: #ff6600; font-weight: bold">
راه‌حل با برنامه‌ریزی خطی
    </div>
    در این مدل سعی می‌کنیم الگوریتمی تقریبی ارائه دهیم که به ما تنها یک عدد می‌دهد و به ما چینش مناسبی ارائه نمی‌کند اما این عدد به ما حدودی از جواب بهینه را می‌دهد.
    <br/>
    ابتدا یک مدل
    IP
    برای این مسئله ارائه می‌دهیم که به صورت زیر است:
    <br/>
    $$x_i \in \{0, p_i\}$$
    $$Constraint:~~\sum_{1 \le i \le n}x_i \le B$$
    $$Maximize:~~\sum_{1 \le i \le n} x_i \times s_i$$
    <br/>
    یکی از ایده‌های روتین برای به‌دست آوردن الگوریتم تقریبی به اصطلاح 
    relax
    کردن IP
    است که به فرم زیر که به فرم 
    LP
    است می‌رسیم:
    $$Constraint:~~\sum_{1 \le i \le n}x_i \le B,~\forall_{1 \le i \le n} 0 \le x_i \le p_i$$
    $$Maximize:~~\sum_{1 \le i \le n} x_i \times s_i$$
<br/>
    می‌دانیم این مسئله راه‌حل چند‌جمله‌ای مثلا با استفاده از الگوریتم
    Simplex
    دارد لذا جوابی به ما می‌دهد آیا این جواب خوب است؟
    <br/>
    ابتدا برای سادگی فرض کنید 
    $s_i$
    ها متمایزند. البته می‌توان به سادگی کاری کرد که در حالت کلی هم این فرض برقرار نباشد اما آن‌را به خواننده واگذار می‌کنیم.
    <br/>
    حالا ادعا می‌کنیم در جواب این مسئله
    LP
    همه
    $x_i$
    ها یا صفرند یا
    $p_i$
    بجز حداکثر یکی از آن‌ها!
    <br/>
    برای اثبات این‌ادعا فرض کنید 
    $i, j$
    داریم که
    $0 < x_i < p_i , 0 < x_j < p_j$
    بدون کاستن از کلیت فرض کنید 
    $s_i > s_j$
    در اینصورت کافیست مقدار
    $x_i$
    را به اندازه
    $\epsilon$
    کم کنیم و 
    $x_j$
    را به همان اندازه اضافه کنیم و این اپسیلون به حد کافی کوچک فرض شده‌است.
    با انجام این‌کار عبارت
    $sum_{1 \le i \le n}x_i \times s_i$
    بیشتر می‌شود که با فرض بیشینه بودن آن در تناقض است.
    <br/>
    این عنصر خاص که 
    $x_k \neq p_k , 0$
    است را عنصر
    $k$
    ام در نظر بگیرید.
    <br/>
    فرض اشیا با اندازه بزرگ را از ابتدا دور ریخته باشیم یعنی 
    $i$
    هاییکه
    $p_i > B$
    است را ابتدای کار دور انداخته‌باشیم.
     حالا ادعا می‌کنیم اگر 
    $i$
    هاییکه
    $x_i = p_i$
    هستند را برداریم
    یک جواب تقریبی خوبی ارائه کرده‌ایم.
    <br/>
    حالا لم زیر را ثابت می‌کنیم:
    <br/>
    <span style="color: red">      
    لم:
        </span>
    اگر جواب به‌دست آمده برابر $V$
    باشد و جواب بهینه برابر
    $OPT$
    باشد داریم:
    $OPT \le V \le 2\times OPT$
    <br/>
    اثبات:
    می‌دانیم در حالت 
    relaxation
    جواب محاسبه شده حتما بهتر از جواب بهینه‌است لذا
    $OPT \le V$
    <br/>
    حالا داریم:
    $$V = \sum_{i : x_i = p_i} x_i \times s_i + x_k \times s_k \le (\sum_{i : x_i = p_i} x_i \times s_i) + (p_k \times s_k)$$
    می‌دانیم
    $p_k \times s_k \le OPT$
    زیرا خود عنصر $k$ام
    به تنهایی یک کاندید برای جواب است و داریم
    $p_k \le B$
    و همچنین بخش
    $\sum_{i : x_i = p_i} x_i \times s_i$
    هم کاندیدی برای جواب است پس:
    $$V \le \sum_{i : x_i = p_i} x_i \times s_i + p_k \times s_k \le 2\times OPT$$
    <br>
    بنابراین اگر 
    $\frac{V}{2}$
    را ارائه کنیم یک ۲تقریب به ما می‌دهد:
    $$\frac{1}{2}OPT \le \frac{V}{2} \le OPT$$

<div id="section2.5" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 17px; color: #ff6600; font-weight: bold">
تقریب $1-\epsilon $
    </div> 
    اکنون برمی‌گردیم به راه‌حل اولی که داشتیم و با استفاده از برنامه‌ریزی پویا مسئله را حل می‌کردیم و سعی می‌کنیم تا به‌وسیله آن ایده راه‌حلی برای یک تقریب بهتر ارائه کنیم.
    الگوریتم ارائه شده 
    $O(n \times p_{max})$
    بود. توجه کنید که ما پارامتری را با این الگوریتم می‌توانیم
    relax
    کنیم پارامتر
    $p$
    یعنی اندازه اشیا است در حالیکه می‌خواهیم تقریبی روی بیشینه ارزش انجام دهیم. به همین خاطر دنبال این هستیم که آیا می‌توانیم الگوریتمی ارائه دهیم که به‌جای اینکه
    $Poly(n, p_i)$
    باشد
    $Poly(n, s_i)$
    باشد. در ادامه چنین چیزی نشان می‌دهیم:

<div/>

<div id="section2.5.1" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 17px; color: red; font-weight: bold">
روشی دیگر برای حل کوله‌پشتی با برنامه‌ریزی پویا
    </div> 
    مسئله کوله‌پشتی‌مان را به مدل دیگری از کوله‌پشتی کاهش می‌دهیم فرض کنید 
    $S = \sum_{i \in [n]} s_i$
    و حالا مسئله جدیدمان یک مسئله کوله‌پشتی است با این تفاوت که شئ 
    $i$ام
    اندازه‌اش برابر
    $s_i$
    و ارزشش برابر 
    $-p_i$
    است.
    <br\>
        حالا فرض کنید در مسئله جدید
        $B = S$
        است و ما الگوریتم پویایی که قبلا به آن اشاره کردیم را روی مسئله اجرا می‌کنیم با این تفاوت که
        $O(i, j)$
        به معنای پر کردن کوله‌پشتی با بیشترین ارزش از اشیا
        ۱ تا $i$
        است به طوریکه جمع اندازه آن‌ها برابر
        «دقیقاً»
        $j$
        بشود.
        تنها کاری که کافیست بکنیم این است که پایه را تغییر دهیم یعنی
        $$O(0, 0) = 0, \forall_{i > 0} O(0, i) = \infty$$
        <br\>
            حالا فرض کنید الگوریتم را اجرا کردیم و به ازای هر 
            $0 \le i \le S$
            مقدار
            $O(n,i)$
            را بررسی می‌کنیم.
            این مقدار جمع یکسری
            $-p_i$
            ها‌را نشان می‌دهد
            $$O(n, i) = \sum_{k \in objects(O(n, i))} -p_k$$
            اگر مقدار بالا بیشتر مساوی 
            $-B$
            بود یعنی حداقل اندازه‌ای که بتواند ارزش
            $i$
            را بسازد کمتر مساوی 
            $B$
            است. (به این جمله خوب فکر کنید!)
            <br/>
            بنابراین هر $i$ ای که این شرط را داشت یعنی
            $O(n, i) \ge -B$
            یعنی می‌توانیم ارزش $i$ را استخراج کنیم با حفظ کردن شرایط مسئله
            کافیست با یک پیمایش آخر به اولی روی 
            $i$
            بزرگترین مقداری را پیدا کنیم که
            $O(n, i) \ge -B$
            در اینصورت جواب مسئله
            «اصلی»
            ما برابر
            $i$
            می‌شود!
            <br\>
                حالا می‌دانیم در روش قبلی به اندازه
                $O(np_max)$
                هزینه می‌کردیم. در این روش باید همه
                $O(i, j), 0 \le i \le n, 0 \le j \le S$
                را بسازیم یعنی
                $O(nS) < O(n\times ns_{max})$
                بنابراین الگوریتمی
                $poly(n, s_{max})$
                به دست آورده‌ایم.
                $s_{max} = max_{i \in [n]} s_i$
                
<div/>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: red; font-weight: bold">
    ادامه کار...
    </div>  
    حالا که الگوریتمی از
    $poly(n, s_{max})$
    داریم کافیست مقدار
    $\epsilon$
    را به عنوان دقت وارد کار کنیم.
   یک پارامتر
    $k = \lfloor \frac{n}{\epsilon} \rfloor$
    تعریف می‌کنیم و سپس ارزش شئ $i$ ام را با مقدار زیر جایگذاری می‌کنیم:
    $$\lfloor \frac{s_i}{s_{max}} \times k \rfloor$$
    و این مسئله را با روشی که ارائه کردیم حل می‌کنیم.
    <br/>
    الان ارزش اشیا در بازه صفر تا
    $\frac{n}{\epsilon}$
    است که نتیجه می‌دهد الگوریتم قبلی ما در
    $O(n^2 \times \frac{n}{\epsilon}$
    عمل می‌کند و براساس پارامتر دقت و ورودی چند جمله‌ای است.
    اگر بخواهیم دقت را افزایش دهیم باید 
    $\epsilon$
    را کم کنیم و در نتیجه
    $\frac{n^3}{\epsilon}$
    زیاد می‌شود.
    <br/>
    دقت بیشتر یعنی هزینه زمانی بیشتر اما نکته‌ای که وجود دارد این است که می‌توانیم به صورت دلخواه آن‌را نتظیم کنیم.
    <br/>
    الگوریتم FPTAS for KnapSack(fully polynomial time approximation scheme) به این ترتیی عمل می کند که توسط kim و Ibarra ارائه شد.
    <br/>
    <div style="font-size: 17px; color: blue; font-weight: bold">
    اثبات این الگوریتم به شکل زیر است:
    </div> 
    فرض کنید
    $\hat{s_i}$
    وزن‌های جدید ما هستند.
    و $OPT$
    جواب بهینه‌است و 
    $\hat{S}$
    زیر مجموعه‌ای از عناصر هست که ما به‌دست می‌آوریم.
    <br/>
    $$\sum_{i \in \hat{S}} s_i \ge \sum_{i \in \hat{S}}(\lfloor \frac{s_i}{s_{max}} k \rfloor \frac{s_{max}}{k})$$
    $$\sum_{i \in \hat{S}} s_i \ge (\frac{s_{max}}{k})\sum_{i \in \hat{S}} \hat{s_i}$$
    مقدار بالا ارزشی است که از الگوریتم
    FPTAS
    به دست می‌آید.
    <br/>
    در ادامه ارزش الگوریتم
    $OPT$
    را بیان می‌کنیم:
    <br/>
    $$(\frac{s_{max}}{k}) \sum_{i \in OPT} \lfloor \frac{s_i}{s_{max}} k \rfloor \ge (\frac{s_{max}}{k} (\sum_{i \in OPT} \frac{s_i}{s_{max}}k - 1)$$
    $$(\frac{s_{max}}{k}) \sum_{i \in OPT} \lfloor \frac{s_i}{s_{max}} k \rfloor \ge \sum_{i \in OPT} s_i - \sum_{i \in OPT} \frac{s_{max}}{k}$$
    $$(\frac{s_{max}}{k}) \sum_{i \in OPT} \lfloor \frac{s_i}{s_{max}} k \rfloor \ge \sum_{i \in OPT} s_i - \frac{ns_{max}}{k}$$
    $$(\frac{s_{max}}{k}) \sum_{i \in OPT} \lfloor \frac{s_i}{s_{max}} k \rfloor \ge \sum_{i \in OPT} s_i - \epsilon s_{max}$$
می‌دانیم که 
    $\epsilon s_{max} \le \epsilon OPT$
    که در نهایت داریم:
    $$(\frac{s_{max}}{k}) \sum_{i \in OPT} \lfloor \frac{s_i}{s_{max}} k \rfloor \ge (1 - \epsilon) OPT$$
<br/>

<div id="section3" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
کاربرد روش‌های تقریبی در بهینه‌سازی گسسته
(مطالعه بیشتر)
    </div>
    فرض کنید می‌خواهیم جواب دقیق مسئله کوله‌پشتی را با استفاده از روش 
    backtracking
    به دست بیاوریم به این‌صورت که در مرحله $i$ام
    به ازای دو حالت انتخاب کردن و نکردن شئ متناظر دو تا انشعاب به درخت 
    backtracking
    می‌دهیم. چنین کاری به صورت نمایی زمان می‌برد و اصلا کارآمد نیست اما با استفاده از روش‌های تقریبی‌ای که در این بخش آموختیم نشان می‌دهیم روند را خیلی بهتر می‌توان کرد.
    <br>
    در این حوزه‌از 
    "Heuristic"
    استفاده می‌کنیم
فرض کنید می‌خواهیم پیش‌بینی‌ای انجام بدهیم که با اشیائی که تا الآن انتخاب کرده‌ایم بهینه‌ترین جواب ممکن با انتخاب بعدی‌ها چه
    «کرانی»
    دارد.
    توجه کنید به دست آوردن جواب دقیق به معنای پیمایش کل زیردرخت متناظر است اما پیدا کردن کران هزینه کمتری می‌برد.
    <br/>
    فرض کنید در رأس $v$
    درخت پسگرد
    (یا همان
    backtracking)
    هستیم و می‌خواهیم ببینیم جواب‌هاییکه با گسترش این جواب به دست می‌آیند حداکثر چند است.
    با استفاده از الگوریتمی نسبتا سریع و با انتخاب یک 
    approximation
    مناسب می‌توانیم این حدود را به‌دست آوریم و اگر مقدار به‌دست آمده که مقداری
    «خوش‌بینانه»
    است بدتر از جوابی بود که تا الآن در درخت پسگرد دیده‌ایم
    می‌توانی به‌طور کلی زیر درخت $v$
    را پیمایش نکنیم!
    <br/>
    این حرکت حرکت مناسبی در روش‌های بهینه‌سازی است که به آن
    branch and bound
    نیز می‌گویند.
    برای مثال اگر از یک
    $\alpha-approximation$
    استفاده کنیم. آنوقت رأس‌هایی از درخت که مقدارشان
    بهتر مساوی
    $\alpha OPT$
    است را می‌بینیم.
   <br/>
    در ادامه کدی پیاده‌سازی شده‌ که از الگوریتم 
    redux
    برای بهینه‌سازی درخت پسگرد در‌آن استفاده شده‌است:

```python

// In the name of God

#include <bits/stdc++.h>

using namespace std;
const int MN = 1e4 + 10;

struct Obj{
	int s, p;
}obj[MN];

bool cmp(const Obj &a, const Obj &b) {
	return 1ll * a.s * b.p > 1ll * b.s * a.p;
}

// return the approximation for objects in interval [l, r)
long long redux_heuristic(int l, int r, int B) { 
	long long return_value = 0;
	long long sum_of_s = 0;
	int remaining_space = B;
	for (int i = l; i < r; i++) {
		remaining_space -= obj[i].p;
		if (remaining_space < 0) {
			return_value = max((long long)obj[i].s, sum_of_s);
			break;
		}
		sum_of_s += obj[i].s;
	}
	return max(return_value, sum_of_s);
}

// find the exact solution for [l, r) with given remaining space
// and given accumulated 's' values from before 
// up until now we have seen a solution with value = 'best_solution'
const long long INF = 1e18;
long long find_solution(int l, int r, long long accumulated_s, int remaining_space, long long best_solution = 0) {
	if (l == r)
		return accumulated_s;
	// prune current node and subtree if it is too small even with optimistic heuristic
	if (redux_heuristic(l, r, remaining_space) + accumulated_s < best_solution)
		return accumulated_s;
	best_solution = max(best_solution, find_solution(l+1, r, accumulated_s, remaining_space, best_solution));
	if (obj[l].p <= remaining_space)
		best_solution = max(best_solution, find_solution(l+1, r, accumulated_s + obj[l].s, remaining_space - obj[l].p, best_solution));
	return best_solution;
}

int main() {
	int n, B;
	cin >> n >> B;
	for (int i = 0; i < n; i++) 
		cin >> obj[i].p >> obj[i].s;
	// exclude all with size larger than B
	for (int i = 0; i < n; i++) {
		if (obj[i].p > B) {
			swap(obj[i], obj[n-1]);
			n--;
		}
	}
	// sort object in decending order of: s[i]/p[i]
	sort(obj, obj + n, cmp);
	
	cout << find_solution(0, n, 0, B) << endl;
	return 0;
}
```
