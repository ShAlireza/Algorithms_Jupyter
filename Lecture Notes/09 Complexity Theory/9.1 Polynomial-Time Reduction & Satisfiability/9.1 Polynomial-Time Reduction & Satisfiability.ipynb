مقدمه</div><br>
    در نظریهٔ محاسبات و نظریهٔ پیچیدگی محاسباتی، کاهش، الگوریتمی برای انتقال یک مسئله به مسئلهٔ دیگری می‌باشد. کاهش از یک مسئله به مسئلهٔ دیگر ممکن است برای نشان دادن اینکه مسئلهٔ دوم حداقل به اندازهٔ مسئلهٔ اول، سخت است استفاده شود. ساختار ریاضی ایجاد شده درمجموعه‌ای از مسائل با کاهش یک نوع خاص، عموماً یک مرتبهٔ پایین‌تر را تشکیل می‌دهد که کلاسهای هم‌ارزی آن ممکن است برای تعریف درجهٔ غیرقابل حل بودن و کلاسهای پیچیدگی استفاده شود. به‌طور شهودی، مسئلهٔ A قابل کاهش به مسئلهٔ B است اگر الگوریتم کارای حل مسئله ی B بتواند به عنوان یک زیر روال برای حل کارای مسئلهٔ A نیز استفاده شود. اگر این اتفاق بیفتد، حل A نمی‌تواند از حل B مشکلتر باشد.

کاهش چند جمله ای
</div><br>
الگوریتم هایی که تا کنون دیده اید راه هایی کارامد برای مسائل ارائه می دادند؛ اما همیشه اینطور نیست و بعضی مسائل از این دست الگوریتم ها ندارند.منظور ما در اینجا از واژه ی کارامد، مفهوم عام آن نیست. شاید از نظر شما الگوریتمی که اردر زمانی $n^{10}$ دارد، الگوریتم کارامدی نباشد. اما در دسته بندی که در ادامه درس آن را به طور دقیقتر یاد خواهید گرفت، این الگوریتم نیز کارامد خواهد بود. در واقع در این دسته بندی اگر الگوریتمی از O(P(n)) باشد که P(n) به معنای عبارتی چند جمله ای بر حسب n است، الگوریتمی کارامد حساب می شود و به مسائلی که با این دست الگوریتم ها حل میشوند، مسائل نرم(tactable)  می گویند. این نوع دسته بندی در نهایت منتهی خواهد شد به دو دسته مسائل P (دسته ای که با الگوریتم های چند جمله ای حل می شوند) و NP که تعریف دقیقتر آنها را در ادامه درس خواهید دید. قسمتی از علوم کامپیوتر به این می پردازد که چه مسائلی در دسته P قرار میگیرند. راهی که میتوان برای پیدا کردن این دست مسائل ارائه داد این است که یک مسئله با چنین الگوریتمی پیدا کنیم و به طریقی یک مسئله جدید را به مسئله قبلی و الگوریتم چند جمله ای آن مربوط کنیم.
<br>
دلیل اینکه این بحث در این قسمت انجام شد این است که با انگیزه ی مطرح شدن مبحث این جلسه آشنا شده باشید که آن هم ارائه راهی است برای ایجاد ارتباط بین مسائلی که در پاراگراف قبل به آنها اشاره کردیم.

مسائل تصمیمی
</div>            
در این مبحث مسائلی را در نظر میگیریم که از نوع تصمیمی(decision problem) هستند.این دسته از مسائل به مسائلی گفته می شوند که پاسخ به آنها با آری یا نه صورت میگیرد.همچنین میتوان مسائلی را که از نوع تصمیمی نیستند، به این نوع مسائل تبدیل کرد.
<br>
برای مثال مسئله ی بیشترین درجه ی یک گراف را میتوان به صورت تصمیمی به این شکل بیان کرد که:
<br>
آیا راسی با درجه بیشتر از k وجود دارد؟

<div style="font-size: 17px; color: #ff6600; font-weight: bold">
مسائل تشخیص زبان
</div>            
مسائل تصمیمی را می توان به شکل مسائل تشخیص زبان$(language-recognition-problem)$  مطرح کرد. فرض کنید U مجموعه تمامی ورودی های ممکن برای مسئله باشد و $L \subset U$ مجموعه تمامی ورودی هایی باشد که جواب مسئله به آنها "بله" است.  حال مسئله به این تبدیل می شود که بفهمیم ورودی عضو L هست یا نه. به L زبان مسئله میگوییم.
<br>
حال فرض کنید L1 و L2 زبان دو فضای U1 و U2 باشند. در این صورت می گوییم L1 به L2 در زمان چندجمله ای کاهش پذیر است وقتی که: الگوریتمی با زمان چندجمله ای موجود باشد که هر ورودی $u1 \subset U1$ را به ورودی $u2 \subset U2$ تبدیل کند به گونه ای که $u1 \subset L1$ باشد اگر و تنها اگر $u2 \subset L2$  باشد.  الگوریتم بر اساس اندازه u1 چندجمله ای است. فرض میکنیم که تعریف اندازه در U1 و U2 خوش تعریف است. پس u2 نیز براساس اندازه u1 چندجمله ای است.
<br>
اگر الگوریتمی در زمان چندجمله ای برای L2 داشته باشیم و L1 نیز به L2 در زمان چندجمله ای کاهش پذیر باشد، آنگاه می توانیم هر ورودی L1 را در زمان چندجمله ای به ورودی L2 تبدیل کنیم و چون الگوریتم چند جمله ای برای حل L2 داریم، آنرا در زمان چند جمله ای حل کنیم. پس با استفاده از دو الگوریتم زمان چندجمله ای می توانیم L1 را حل کنیم که در مجموع الگوریتمی چند جمله ای است.

خواص کاهش چندجمله ای
</div>            
مفهوم کاهش چندجمله ای دوطرفه نیست. یعنی ممکن است L1 به L2 در زمان چندجمله ای کاهش یابد اما L2 به L1 کاهش نیابد. اگر هم L1 به L2 در زمان چندجمله ای کاهش یابد و هم L2 به L1 ، آنگاه میگوییم L1 و L2 به صورت چندجمله ای، معادل هستند.
<br>
خاصیت تعدی در مفهوم کاهش چندجمله ای وجود دارد. به این معنی که اگر L1 به L2 در زمان چند جمله ای کاهش پذیر باشد و L2 نیز در زمان چندجمله ای به L3 کاهش پذیر باشد، آنگاه L1 به L3 نیز در زمان چند جمله ای کاهش پذیر است؛ زیرا ورودی های L1 در زمان چند جمله ای به ورودی های L2 تبدیل می شوند و ورودی های L2 نیز در زمان چندجمله ای به ورودی های L3 تبدیل می شوند. پس با اجرای دو الگوریتم چند جمله ای می توان ورودی های L1 را به ورودی های L3 تبدیل کرد و جمع این دو الگوریتم چند جمله ای نیز ، الگوریتمی چند جمله ای بدست می دهد.
<br>
اصولا به این مفهوم در این قسمت از درس به این منظور پرداخته شده است که وقتی نمی توانیم مسئله ای را با الگوریتمی کارامد حل کنیم، به دنبال مسئله ای به صورت چندجمله ای معادل می گردیم که آن مسئله مسئله ای سخت باشد.آنگاه می توان مسئله جدید را در دسته بندی مسائل سخت قرار داد.

روش کار</div>
    ایده ی نشان دادن اینکه یک مسئله سخت‌تر یا آسان‌تر از مسئله‌ی دیگری نیست حتی زمانی که هر دو مسئله‌، مسئله‌ی تصمیم‌گیری باشند نیز به کار می‌رود. ما از این ایده تقریباً برای اثبات تمامی مسائل NPC استفاده می‌کنیم.
اجازه دهید یک مسئله‌ی تصمیم‌گیری مانند A را در نظر بگیریم که می‌خواهیم آن را در پیچیدگی زمانی چندجمله‌ای O(nk) حل کنیم. ورودی را به یک مسئله‌ی خاص به‌عنوان «نمونه‌ای» از این مسئله فراخوانی می‌کنیم. به‌عنوان‌مثال در مسئله‌ی مسیر‌، یک نمونه می‌تواند یک گراف خاص G باشد با دو رأس خاص u و v از G و مقداری مشخص برای پارامتر k. حال فرض می‌کنیم که یک مسئله‌ی تصمیم‌گیری دیگری داریم با نام B‌ و میدانیم که این مسئله راه‌حلی در پیچیدگی زمانی چندجمله‌ای دارد. درنهایت فرض می‌کنیم رویه‌ای داریم که هر یک از نمونه‌های a از مسئله‌ی A را به نمونه‌های b از مسئله‌ی B با خصوصیات زیر تبدیل می‌کند یا «کاهش می‌دهد».
    <br>
    <ol>
        <li>عملیات کاهش دارای پیچیدگی زمانی چندجمله‌ای است.</li>
        <li>جواب‌های هر دو مسئله یکسان است؛ یعنی جواب مسئله‌ی a «بلی» است اگر و فقط اگر جواب مسئله‌ی b «بلی» باشد.</li>
    </ol>
    پس کلاً سه مرحله‌داریم:

<br>
    <ol>
        <li>نمونه‌ی a از مسئله‌ی A را با استفاده از «الگوریتم کاهشی» با پیچیدگی زمانی چندجمله‌ای به نمونه‌ی b از مسئله‌ی B تغییر شکل می‌دهیم.</li>
        <li>الگوریتم کاهشی را بر روی نمونه‌ی b از مسئله B اجرا می‌کنیم.</li>
        <li>از جوابی که برای b به دست آورده‌ایم به‌عنوان جوابی برای مسئله‌ی a استفاده می‌کنیم.</li>
    </ol>
