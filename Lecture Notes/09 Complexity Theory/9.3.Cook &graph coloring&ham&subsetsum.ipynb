<br>
بعضی وقتها برای حل کردن یک مسئله راه های زیادی وجود دارد.
<br>
اول از همه برای خود تعریف کنیم معنی حل مسئله چیست؟ رسیدن به جواب نهایی با حالت اپتیمال حل مسئله نامیده میشود.
اکنون که تعریف درستی از حل مسئله داریم سوال دیگری مطرح میشود، اگر چند راه حل داشته باشیم که همه ی ان ها مسئله-
ی مارا حل کنند کدام راه حل از همه بهتر است؟ .
<br>
جواب مشخص است راه حلی که زودتر از همه مارا به جواب برساند.
دانشمندان علوم کامپیوتر و ریاضی برای محاسبه و مقایسه سرعت راه حل ها از تعریفی به نام پیچیدگی زمانی استفاده میکنند.
<br>
پیچیدگی زمانی حداقل زمان مورد نیاز برای حل مسئله با توجه به اندازهی ورودی های مسئله است.

پیچیدگی زمانی
</div><br>
فرض کنید به خانه ی اخر یک ارایه ی خالی میخواهید عددی را اضافه کنید شما تنها لازم است درون خانه ی اخر ارایه یک عدد
بریزید بنابر این این کار را با پیچیدگی زمانی ثابت 1 انجام میدهید که ان را به صورت:
<br>
$
𝑂(1)
$
<br>
نشان میدهند.
<br>
حال فرض کنید که میخواهید تعداد n عدد را با هم جمع بزنید رویکرد شما این است که از ابتدا شروع کرده و اعداد را با هم
جمع میزنید باید برای رسیدن به هدف خود تعداد n جمع انجام دهیم پس پیچیدگی زمانی ما n است:
<br>
$
𝑂(𝑛)
$
<br>
دیدیم که هردوی این مسائل در زمان چند جملهای حل میشوند که به آن ها مساثل polynomial گفته میشود.
<br>
اما مسائلی موجود هستند که در زمان چند جملهای حل نمیشوند برای مثال:
 <br>
فرض کنید تعدادی شهر داریم و هزینه رفتن مستقیم از یکی به دیگری را میدانیم. مطلوب است کمهزینهترین مسیری که از یک
شهر شروع شود و از تمامی شهرها دقیقاً یکبار عبور کند و به شهر شروع بازگردد .
<br>
برای محاسبهی این مقدار باید تمام n! جایگشت شهرها را بدست آوریم.
 <br>
محاسبه ی این مقدار یک غیر جند جمله ای است و پیچیدگی زمانی این مسئله با تعداد شهر n برابر:
<br>
$
𝑂(𝑛!)
$
<br>
است، به این سری مسائل که در زمان چند جمله ای حل نمیشوند مسائل non polynomial گفته میشود.
برخی از مسائل NP هستند که اگر جواب انها را داشته باشیم میتوانیم درستی جواب را در زمان چند جملهای پیدا کنیم به این
مسائل، مسائل NP-complete گفته میشود.

</div><br>

باشد؟
    <br>
سوال را برای مثال زیر از حل کنید :
    <br>
𝐴∧𝐵
    <br>
جواب سوال بله است اگر ارزش هر دو متغیر درست باشد کل گزارهی ما درست است اما پیچیدگی زمانی این مسئله چیست ؟

</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
استفن کوک در سال 1971 اثبات کرد مسئلهی SAT اولین مسئله ی NP-Complete شناخته شده است و تا آن زمان مفهومی برای
NP-Complete ساخته نشده بود تئوری کوک این مسئله را مطرح کرد که:
<br>
هر مسئله ی NP میتواند در زمان چند جمله ای به یک مسئلهی SAT تبدیل شود
<br>
به عبارت دیگر برای هر مسئلهی تصمیم گیری یک ماشین تورینگ غیرقطعی میسازیم که بتواند مسئله را در زمان چند جمله ای حل کند
سپس ورودی ماشین را با یک عبارت دودویی مدل میکنیم که به ما بگوید آیا با این ورودیها مقدار خروجی نهایی بله T  است .
پس این ماشین کاهش داده شده به مسئله ی SAT است که ایا با این ورودی ها گزارهی نهایی T است یا F

دور همیلتونی
</div><br>
<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
</div>     
دور همیلتونی، دوری است که هر راس را دقیقاً یک بار مشاهده می‌کند.(به جز راسی که هم به عنوان آغاز و هم پایان می‌باشد. در نتیجه این راس دو بار دیده می‌شود.) به‌طور قراردادی، گرافی کوچک شامل یک راس، دارای دور همیلتونی است، ولی گراف متصلی دارای دو راس، شامل دور همیلتونی نیست.
<hr>
</div>


<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
</div>     
<br>
مسئله ی دور همیلتونی: گراف غیر جهت دار G = (V,E) داده شده است، آیا یک دور ساده‌ی Γ وجود دارد که شامل همه‌ی راس‌های V باشد؟
<br>
<img src="Untitled Folder/ham1.PNG">
<br>
قضیه : مسئله‌ی دور همیلتونی جهت دار به مسئله‌ی دور همیلتونی کاهش می‌یابد. (dir_ham_cycle $\leqslant_{p}$ ham_cycle) 
<br>
اثبات :
<br>
گراف G با  n راس داده شده است. ما گراف G’ را به 3n راس مانند تصویر می‌سازیم.
<br>
<img src="Untitled Folder/ham2.PNG">
<br>
G دور همیلتونی جهت دار دارد اگر و فقط اگر ${G}'$ دور همیلتونی داشته باشد. 
<br>
$\Leftarrow $ فرض کنید G دور همیلتونی جهتدار دارد، آنگاه ${G}'$ هم دور همیلتونی‌ای با همان ترتیب دارد.
<br>
$\rightarrow $ 
فرض کنید ${G}'$ دور همیلتونی غیر جهت دار ${Γ}'$ دارد. آنگاه ${Γ}'$ باید راس های ${G}'$ را با یکی از دو ترتیب زیر مشاهده کند.
<br>
…B,G,R,B,G,R,…
<br>
…B,R,G,B,R,G…
<br>
ترتیب راس های آبی یا برعکس این ترتیب در ${Γ}'$ یک دور همیلتونی جهت دار $Γ$ در $G$ می‌سازد.
<br>
قضیه: مسئله‌ی 3sat به مسئله‌ی دور همیلتونی جهتدار کاهش می‌یابد. (3sat $\leqslant_{p}$ dir_ham_cycle)
<br>
اثبات: فرض کنید ما یک 3sat I با متغییر های $x_{1},x_{2},…,x_{n}$ و کلازهای $C_{1},C_{2},…,C_{k}$ داشته باشیم.
<br>
حال گراف G را به صورت زیر می‌سازیم.
<br>
گراف هایی می‌سازیم( که gadget نامیده می‌شوند) که بیانگر متغییر ها باشد.
<br>
گراف هایی می‌‌سازیم که بیانگر کلازها باشند.
<br>
این گراف ها را طوری به هم متصل می‌کنیم که گراف مورد نظر را بسازد.
<br>
سپس نشان می دهیم این گراف دور همیلتونی جهت دار رارد اگر و فقط اگر I را satisfy کند.
<br>
برای هر متغییر gadget ها را به صورت زیر می‌سازیم.
<br>
<img src="Untitled Folder/ham3.PNG">
<br>
که نوع حرکت دور همیلتونی در این gadget بیانگر صحیح یا غلط بودن متغییر $x_{i}$ است. اگر از سمت راست به چپ بود یعنی $x_{i}$ درست است، اگر از سمت چپ به راست بود یعنی $x_{i}$ درست است.
<br>
حال به ازای هر کلاز مانند زیر عمل می‌کنیم. اگر $x_{i}$ در کلاز $c_{j}$ بود آنرا به صورت زیر و اگر $!x_{i}$ در کلاز $c_{k}$ بود آن را مانند زیر متصل می کنیم.
<br>
<img src="Untitled Folder/ham4.PNG">
<br>
حال تمام gadget ها را مانند زیر به یکدیگر متصل می‌کنیم و رو راس s و t را مانند زیر به گراف اضافه می‌کنیم تا گراف G مانند زیر ساخته شود.
<br>
<img src="Untitled Folder/ham5.PNG">
<br>
با توجه به جهت دور همیلتونی در هر  gadget یک دور همیلتونی برای هر متغییر یک مقدار true یا false در نظر می ‌گیرد.
<br>
برای آنکه دور همیلتونی باشد باید تمامی clause ها را ملاقات کنیم.
<br>
ما تنها در صورتی می توانیم یک کلاز را ملاقات کنیم که I را satisfy کنیم.
<br>
بنابراین اگر یک دور همیلتونی وجود داشته باشد، یک جواب satisfy برای I وجود دارد.
<br>
حال به راحتی می‌توان اثبات کرد که dir_ham_cycle در NP است. (چرا؟)

    رنگ آمیزی گراف
</div><br>            
 در تئوری گراف مسئله ای به نام رنگ آمیزی گراف وجود دارد که حالت خاصی از لیبل گذاری گراف است.
    <br> 
    مسئله ی اصلی رنگ امیزی نودهای گراف است با حداقل رنگ به گونهای که هر نود گراف با نودهای مجاورش رنگ مختلف داشته
باشد. این مسئله در این حالت NP-Complete است یعنی در زمان چند جملهی پاسخی برای این مسئله موجود نیست.
    <br> 
    نوع دیگری از این مسئله وجود دارد که با تعداد محدودی رنگ نودهای گراف باید به گونهای رنگ آمیزی شود که هیچ دو نود
مجاوری همرنگ نباشد ابن مسئله نیز از 3 رنگ به بالا NP-Complete است.
    
</div>

```python
#include <iostream> 
#include <cstdlib>
#include <list>
using namespace std; 
```

```python

// A class that represents an undirected graph
class Graph
{
    int V;    // No. of vertices
    list<int> *adj;    // A dynamic array of adjacency lists
public:
    // Constructor and destructor
    Graph(int V)   { this->V = V; adj = new list<int>[V]; }
    ~Graph()       { delete [] adj; }
 
    // function to add an edge to graph
    void addEdge(int v, int w);
 
    // Prints greedy coloring of the vertices
    void greedyColoring();
};
```

```python
void Graph::addEdge(int v, int w)
{
    adj[v].push_back(w);
    adj[w].push_back(v);  // Note: the graph is undirected
}
 
// Assigns colors (starting from 0) to all vertices and prints
// the assignment of colors
  
```

```python

void Graph::greedyColoring()
{
    int result[V];
 
    // Assign the first color to first vertex
    result[0]  = 0;
 
    // Initialize remaining V-1 vertices as unassigned
    for (int u = 1; u < V; u++)
        result[u] = -1;  // no color is assigned to u
 
    // A temporary array to store the available colors. True
    // value of available[cr] would mean that the color cr is
    // assigned to one of its adjacent vertices
    bool available[V];
    for (int cr = 0; cr < V; cr++)
        available[cr] = false;
 
    // Assign colors to remaining V-1 vertices
    for (int u = 1; u < V; u++)
    {
        // Process all adjacent vertices and flag their colors
        // as unavailable
        list<int>::iterator i;
        for (i = adj[u].begin(); i != adj[u].end(); ++i)
            if (result[*i] != -1)
                available[result[*i]] = true;
 
        // Find the first available color
        int cr;
        for (cr = 0; cr < V; cr++)
            if (available[cr] == false)
                break;
 
        result[u] = cr; // Assign the found color
 
        // Reset the values back to false for the next iteration
        for (i = adj[u].begin(); i != adj[u].end(); ++i)
            if (result[*i] != -1)
                available[result[*i]] = false;
    }
 
    // print the result
    for (int u = 0; u < V; u++)
        cout << "Vertex " << u << " --->  Color "
             << result[u] << endl;
}
```

```python
// Driver program to test above function
int main()
{
    Graph g1(5);
    g1.addEdge(0, 1);
    g1.addEdge(0, 2);
    g1.addEdge(1, 2);
    g1.addEdge(1, 3);
    g1.addEdge(2, 3);
    g1.addEdge(3, 4);
    cout << "Coloring of graph 1 \n";
    g1.greedyColoring();
 
    Graph g2(5);
    g2.addEdge(0, 1);
    g2.addEdge(0, 2);
    g2.addEdge(1, 2);
    g2.addEdge(1, 4);
    g2.addEdge(2, 4);
    g2.addEdge(4, 3);
    cout << "\nColoring of graph 2 \n";
    g2.greedyColoring();
 
    return 0;
```

مسئله ی جمع زیر مجموعه : یک مجموعه از اعداد طبیعی $w_{1},w_{2},w_{3},…,w_{n}$ داده شده است. آیا زیرمجموعه‌ای از این مجموعه وجود دارد که جمع اعداد آن W شود؟
<br>
مثال: 
<br>
EX. {1,3,4,17,22,34,96,115,202} , W = 241
<br>
Yes. {1,4,34,202}

قضیه: مسئله‌ی 3sat به مسئله‌ی جمع زیرمجموعه کاهش می‌یابد. (sat $\leqslant_{p}$ subset sum)
<br>
اثبات:
یک نمونه 3sat با n متغییر و k کلاز به ما داده شده است. 2n+2k عدد میسازیم که دارای n+k رقم هستند:
یک رقم برای هر $x_{i}$ و هر $c_{j}$ وجود دارد.
<br>
دو عدد برای هر متغییر $x_{i}$ وجود دارد.
<br>
دو عدد برای هر کلاز $c_{j}$ وجود دارد.
<br>
جمع هر رقم $x_{i}$ یک و جمع هر رقم $c_{j}$ چهار است.

I satisfiable است اگر و فقط اگر یک زیر مجموعه وجود داشته باشد که جمعش W شود.
<br>
فرض کنید I satisfiable است، آن اعدادی را انتخاب میکنیم که باعث true شدن I می شوند.
<br>
از آنجا که I satisfiable است مجموع ارقام هر ستون $c_{j}$ هداقل یک و حداکثر 3 است.
<br>    
یک مجموع میابیم که مجموع رقم $c_{j}$ را چهار کند.
<br>    
حال فزض می‌کنیم که یک مجموعه از اعداد مجود دارد که مجموعشان W است:
<br>

<br>    
رقم $c_{j}$ مجبورمان می کند که حداقل یک literal در هر کلاز را انتخاب کنیم.
<br>    
$x_{i}$ را true درنظر میگیریم اگر و فقط اگر ردیف $x_{i}$ انتخاب شده باشد.
