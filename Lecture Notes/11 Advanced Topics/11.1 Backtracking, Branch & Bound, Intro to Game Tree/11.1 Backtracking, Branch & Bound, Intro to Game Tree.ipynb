<div style="font-size: 22px; color: #ff6600; font-weight: bold">
توصیف حالت‌های مسئله
</div>

<br>
    هدف از روش پس‌گرد، پیمایش حالت‌های مختلف برای یافتن پاسخ مسئله است. در ادامه با ذکر چند مثال، مفهوم حالت‌های مسئله را مرور می‌کنیم.
    <br> <br>
    <b>
    مثال 1 – مسئله‌ی رنگ‌آمیزی گراف با k رنگ: 
    </b>
    در این مسئله گراف G(V,E) و عدد k داده شده است. می‌خواهیم با استفاده از k رنگ، رأس‌های گراف را رنگ‌آمیزی کنیم؛ به شرطی که دو رأس مجاور، هم‌رنگ نباشند. در صورت ممکن بودن رنگ‌آمیزی، خروجی مسئله «بله» و در غیر این صورت «خیر» است.
    <br>
    «حالت‌»های این مسئله، رنگ‌آمیزی‌های مختلف رئوس گراف هستند. به شکل 1 توجه کنید.
    <br> <br>
    <center>
        <figure>
            <img src="Resources/Figure 1.png", width="65%", style="float:center">
            <br>
            <caption>شکل 1 – سه حالت مختلف از مسئله‌ی رنگ‌آمیزی گراف</caption>            
        </figure>
    </center>
    <br>
    حالت‌های مختلف، ممکن است در شرط مسئله صدق کنند یا نکنند. در این مثال، اگر حالتی پیدا شود که شرط مسئله را برآورده کند، خروجی مسئله «بله» خواهد بود.
<br> <br>
    <b>
    مثال 2 – مسئله‌ی عبور از ماز:  
    </b>
    در این مسئله یک صفحه‌ی شطرنجی با خانه‌های قابل عبور و غیرقابل عبور به ما داده شده است. هم‌چنین دو خانه از صفحه به عنوان نقاط شروع و پایان مشخص شده‌اند. می‌خواهیم از نقطه‌ی شروع به نقطه‌ی پایان برسیم، به شرطی که فقط از خانه‌های قابل عبور رد شویم. در صورت وجود مسیر از نقطه‌ی شروع به نقطه‌ی پایان، خروجی مسئله «بله» و در غیر این صورت «خیر» است.
    <br>
    برای تعریف یک «حالت» در این مسئله، کافی است در هر خانه جهت حرکت بعدی را مشخص کنیم. به شکل 2 توجه کنید.
    <br> <br>
    <center>
        <figure>
            <img src="Resources/Figure 2.png", width="65%", style="float:center">
            <caption>شکل 2 – دو حالت مختلف از مسئله‌ی عبور از ماز</caption>            
        </figure>
    </center>
    <br>
    فقط ترکیب‌هایی را که تشکیل مسیر می‌دهند، به عنوان حالت در نظر می‌گیریم. جهت حرکت در خانه‌هایی که جزء مسیر نیستند مهم نیست.
<br> <br>
    <b>
    فکر کنید:  
    </b>
    برای مسائل زیر، حالت‌های مسئله را شناسایی کنید:
    <br>
    <p dir="ltr">
        N-Queens Problem, Sudoku, Kirkman Schoolgirls Problem
    </p>
    <br>
    <b>
        شبه‌حالت، متغیر و مقدار مجاز
    </b>
    <br>
    برای توصیف هر حالت از تعدادی «متغیر» استفاده می‌کنیم. به عنوان مثال، در مسئله‌ی رنگ‌آمیزی گراف، به ازای هر رأس گراف یک متغیر داریم که می‌تواند مقادیر «قرمز»، «سبز» یا «آبی» را بپذیرد.
    <br>
    به ترکیبی که در آن زیرمجموعه‌ای از متغیرها مقداردهی شده باشند، «شبه‌حالت» گفته می‌شود. شکل 3 یک شبه‌حالت از مسئله‌ی رنگ‌آمیزی گراف را نشان می‌دهد.
    <br> <br>
    <center>
        <figure>
            <img src="Resources/Figure 3.png", width="20%", style="float:center">
            <br>
            <caption>شکل 3 – یک شبه‌حالت مجاز از مسئله‌ی رنگ‌آمیزی گراف</caption>            
        </figure>
    </center>
    <br>
به شبه‌حالتی که شرط مسئله در آن نقض نشده باشد، «شبه‌حالت مجاز» گفته می‌شود. اگر با داشتن یک شبه‌حالت مجاز، یکی از متغیرهای باقی مانده را طوری مقداردهی کنیم که شرط مسئله نقض نشود، یک «شبه‌حالت مجاز» دیگر ایجاد می‌شود. به مقدار تعیین شده برای متغیر، «مقدار مجاز» می‌گوییم.
    <br>
    در ادامه از تعاریف انجام شده برای توضیح روش پس‌گرد استفاده خواهیم کرد. هدف نهایی، یافتن «حالت‌های مجاز» مسئله است.
<br> <br>
    <b>
        نکته:  
    </b>
    انواع مسائل قابل حل با روش پس‌گرد عبارت‌اند از:
    <br>
    <b>
        مسئله‌ی تصمیم‌گیری (Decision Problem):  
    </b>
    بررسی وجود حالت مجاز. پاسخ این مسئله «بله» یا «خیر» است.
    <br>
    <b>
        مسئله‌ی شمارش (Enumeration Problem):
    </b>
    شمارش تعداد حالت‌های مجاز. پاسخ این مسئله یک عدد است.
    <br>
    <b>
        مسئله‌ی بهینه‌سازی (Optimization Problem): 
    </b>
    انتخاب مناسب‌ترین حالت از میان مجموعه‌ی حالت‌ها. پاسخ این مسئله نیز یک عدد است.
    <br>
    البته در هر کدام از این مسائل، می‌توانیم حالت(های) متناظر با جواب را نیز به دست آوریم.
<br> <br>
    <b>
        نکته: 
    </b>
    روش پس‌گرد برای حل مسائلی مناسب است که دارای تعداد حالت‌های شمارا هستند.
</div>


<hr>

<div id="section2" align="justify" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 22px; color: #ff6600; font-weight: bold">
        روش پس‌گرد
    </div>
    <br>
    <b>
    درخت پس‌گرد (Backtrack Tree): 
    </b>
    برای پیمایش حالت‌های مسئله از یک درخت استفاده می‌کنیم. هر «گره» از این درخت نشان دهنده‌ی یک شبه‌حالت مجاز برای مسئله است. هم‌چنین هر «شاخه» از درخت، مشخص شدن مقدار یک (یا چند) متغیر را نشان می‌دهد. ریشه‌ی درخت، شبه‌حالتی است که در آن مقدار هیچ متغیری مشخص نشده باشد.
    <br> <br>
در روش پس‌گرد، درخت پس‌گرد هم‌زمان با پیمایش ساخته می‌شود. از ریشه‌ی درخت شروع می‌کنیم و در هر مرحله با داشتن یک شبه حالت:
    <br>
    1) یکی از متغیرهای باقی مانده را انتخاب می‌کنیم،
    <br>
    2) به ازای تمامی مقادیر مجاز متغیر، یک شاخه به درخت اضافه می‌کنیم (انشعاب یا Branch)،
    <br>
    3) یکی از شاخه‌ها را انتخاب کرده و با داشتن گرهِ فرزند، وارد مرحله‌ی بعد می‌شویم.
    <br> <br>
    اگر به مرحله‌ای برسیم که برای یکی از متغیرهای باقی مانده هیچ مقدار مجازی موجود نباشد، روی درخت به عقب برمی‌گردیم (پس‌گرد یا Backtracking) تا به اولین مجموعه از شاخه‌های انتخاب نشده برسیم. سپس یکی از آن‌ها را انتخاب می‌کنیم.
    <br>
    این کار را تا جایی انجام می‌دهیم که یک حالت مجاز تولید شود. شکل 4، درخت پس‌گرد را برای مسئله‌ی رنگ‌آمیزی گراف نشان می‌دهد.
    <br> <br>
    <center>
        <figure>
            <img src="Resources/Figure 4.png", width="65%", style="float:center">
            <br>
            <caption>شکل 4 – درخت پس‌گرد برای حل مسئله‌ی رنگ‌آمیزی گراف با 3 رنگ (پیمایش اولیه به سمت چپ انجام شده است.)</caption>            
        </figure>
    </center>
    <br> <br>
    <b>
        نکته: 
    </b>
    برای تفکیک «گره»های درخت از «رأس»های گراف، آن‌ها را با مربع نشان می‌دهیم. هم‌چنین به تفاوت کاربرد کلمات «شاخه» و «یال» توجه کنید.
    <br>
    <b>
        نکته: 
    </b>
    تا این لحظه در مورد «نحوه‌ی انتخاب متغیر باقی مانده» و «نحوه‌ی انتخاب مقدار مجاز» صحبتی نکرده‌ایم. اما به نظر می‌رسد ترتیب انتخاب متغیرها و مقادیر، بر زمان رسیدن به حالت مجاز، مؤثر باشد.
    <br> <br>
    قطعه کد زیر، حل مسئله‌ی رنگ‌آمیزی گراف با 4 رنگ را نشان می‌دهد.
</div>

<div id="section3" align="justify" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 22px; color: #ff6600; font-weight: bold">
    تکنیک انشعاب و حد
    </div>
    <br>
    تکنیک انشعاب و حد، برای حل مسائل بهینه‌سازی به روش پس‌گرد استفاده می‌شود. در این مسائل، یافتن یک حالت مجاز کافی نیست، بلکه می‌خواهیم حالت مجازی را پیدا کنیم که یک مقدار خاص در آن بهینه باشد.
    <br> <br>
    <b>
    مثال 1 – مسئله‌ی رنگ‌آمیزی گراف با کمترین رنگ: 
    </b>
    در این مسئله گراف G(V,E) داده شده است. می‌خواهیم با استفاده از کمترین تعداد رنگ، رأس‌های گراف را رنگ‌آمیزی کنیم؛ به شرطی که دو رأس مجاور، هم‌رنگ نباشند. خروجی مسئله کمترین تعداد رنگ‌های ممکن است.
    <br> <br>
برای تشکیل و پیمایش درخت پس‌گرد مسئله، به ازای هر گره درخت پس‌گرد، یک متغیر در نظر می‌گیریم. هم‌چنین یک متغیر سراسری (Global) نیز تعریف می‌کنیم. مقدار اولیه‌ی متغیر سراسری را بی‌نهایت در نظر می‌گیریم. متغیر هر گره نشان دهنده‌ی تعداد رنگ‌های استفاده شده تا آن مرحله (حداقل تعداد رنگ‌های مورد نیاز) و متغیر سراسری نشان دهنده‌ی حداکثر تعداد رنگ‌های مورد نیاز است. حال، الگوریتم پس‌گرد را به صورت زیر تغییر می‌دهیم. فرض کنید مقدار متغیر سراسری را با k نشان دهیم.
    <br> <br>
در هر مرحله از پیمایش، اگر در گرهی قرار داشتیم که مقدار متغیر کمتر از k بود، پیمایش را ادامه می‌دهیم تا به یک برگ برسیم. حال، مقدار متغیر برگ را به جای k قرار می‌دهیم (کران‌گذاری یا Bound). سپس، عملیات پس‌گرد را انجام می‌دهیم.
    <br>
اگر در پیمایش درخت پس‌گرد، وارد گرهی شدیم که مقدار متغیر آن بیشتر از k بود، از ادامه‌ی پیمایش صرف نظر می‌کنیم و به روش پس‌گرد، مسیر دیگری را برای پیمایش انتخاب می‌کنیم.
    <br> <br>
می‌توانید حل این مسئله را در قطعه کد زیر ببینید.
</div>

<div align="justify" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <br>
    <b>
        مثال 2 – مسئله‌ی یافتن گروهک بیشین (Maximal Clique Problem): 
    </b>
    <br>
    <b>
        تعریف گروهک: 
    </b>
    در یک گراف، به زیرگرافی که تمام رئوس آن دوبه‌دو مجاور باشند، گروهک گفته می‌شود.
    <br>
    در این مسئله گراف G(V,E) داده شده است و قصد داریم بیشترین اندازه‌ی ممکن برای یک گروهک را بیابیم. این مسئله را به Integer Linear Programming تبدیل می‌کنیم.
    <br> <br>
    به ازای هر رأس $i$، یک متغیر دودویی ($x_i$) در نظر می‌گیریم که حضور یا عدم حضور آن در گروهک را نشان می‌دهد:
    <br>
    \begin{equation}
    0≤x_i≤1
    \end{equation}
    هم‌چنین به‌ازای هر دو رأس غیرمجاور $i$ و $j$ داریم:
    <br>
    \begin{equation}
    x_i+x_j≤1
    \end{equation}
    هدف ما، بیشینه کردن حاصل جمع متغیرها است:
    <br>
    \begin{equation}
    \max⁡    f=x_1+x_2+⋯x_n
    \end{equation}
    $n$ تعداد رأس‌های گراف است.
    <br> <br>
    این مسئله را نیز می‌توانیم به کمک روش پس‌گرد و تکنیک انشعاب و حد حل کنیم. قطعه کد زیر، حل مسئله را نشان می‌دهد.

</div>

<div id="section4" align="justify" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 22px; color: #ff6600; font-weight: bold">
    آشنایی با بازی‌ها و دسته‌بندی آن‌ها
    </div>
    <br>
    در ادامه‌ی این جزوه، قرار است با درخت بازی آشنا شوید. اطلاعات بیشتر و الگوریتم‌های مربوط آن در درس هوش مصنوعی ارائه می‌شوند. در ابتدا کمی راجع به بازی‌ها صحبت می‌کنیم و سپس، با درخت بازی آشنا می‌شوید.
    <br> <br>
روش‌های مختلفی برای دسته‌بندی بازی‌ها وجود دارد. مثلاً از این لحاظ که «آیا فرایندهای تصادفی نقشی در بازی ایفا می‌کنند یا خیر؟» یا «آیا هر بازیکن از کل حالت‌ها و اعمال گذشته و فعلی بازی مطلع است یا فقط بخشی از آن را می‌داند؟» می‌توان بازی‌ها را طبقه‌بندی کرد. هم‌چنین تعداد بازیکنان، مسئله‌ای تعیین کننده است. شکل 5، مثال‌هایی از بازی‌های مختلف را نشان می‌دهد.
    <br> <br>
    <center>
        <figure>
            <img src="Resources/Figure 5.jpg", width="65%", style="float:center">
            <br>
            <caption>شکل 5 – دسته‌بندی بازی‌ها</caption>            
        </figure>
    </center>
    <br>
    در یکی از روش‌هایی که برای توصیف بازی‌های قطعی (Deterministic) وجود دارد. بازی به کمک شش‌تایی زیر توصیف می‌شود:
    <br>
    <ul>
        <li> مجموعه‌ی حالت‌ها ($S$) (شروع با $s_0$) </li>
        <li> مجموعه‌ی بازیکنان ($P$) (معمولاً به صورت نوبتی بازی می‌کنند.) </li>
        <li> مجموعه‌ی کنش‌ها ($A$) (می‌تواند به حالت یا بازیکن بستگی داشته باشد.) </li>
        <li> تابع گذار، که به صورت $S×A→S$ تعریف می‌شود و نشان می‌دهد از هر حالت و با هر کنش، چگونه حالت بازی تغییر می‌کند. </li>
        <li> مجموعه‌ی حالت‌های پایانی ($F$) </li>
        <li> تابع پاداش به ازای هر حالت پایانی که به صورت $F×P→R$ تعریف می‌شود. $R$ مجموعه‌ی پاداش‌های ممکن است. </li>
     </ul>
    <br>
هدف از انجام بازی، کسب بیشترین پاداش ممکن است. «راه حل»، «استراتژی» یا «سیاست» یک بازی، تابعی به صورت $S→A$ است که کنش انتخابی در هر حالت را مشخص می‌کند؛ به طوری که پاداش بازیکن، بیشینه شود.
    


</div>

    آشنایی با درخت بازی
    </div>
    <br>
    برای یافتن راه حل یک بازی، لازم است برای هر حالت، کنشی که منجر به بیشترین پاداش برای بازیکن می‌شود را پیدا کنیم. درخت بازی به ما کمک می‌کند که این کار را انجام دهیم. گره‌های این درخت، حالت‌های بازی هستند. شاخه‌های خارج شده از هر گره، کنش‌های ممکن را نشان می‌دهند و به حالت حاصل ختم می‌شوند. برگ‌های این درخت، حالت‌های پایانی هستند. با تحلیل این درخت، می‌توان استراتژی مربوط به هر حالت را پیدا کرد. ادامه‌ی این مبحث در درس هوش مصنوعی ارائه می‌شود. در ادامه با دو نمونه از درخت بازی آشنا خواهید شد.
