مسئله‌ی اول: تطابق در گراف 2 بخشی
</div>
مساله اول پیدا کردن بیشترین تطابق در یک گراف 2 بخشی داده شده است . گراف 2 بخشی گرافی است که بتوان راس های آن را به دو بخش تقسیم کرد 
به طوریکه هر راس با راس های بخش خود هیچ یالی نداشته باشد .
حال میخواهیم در چنین گرافی بیشتری تطابق را پیدا کنیم . تطابق زیر گرافی است که در آن هر راس حداکثر درجه 1 داشته باشد یا به عبارتی دیگر هیچ دو یالی وجود نداشته باشد 
که راسی را به اشتراک بگذارند

 راه حل
</div>
برای حل مساله با استفاده از مساله بیشینه شار گراف G
داده شده را به گراف H تبدیل میکنیم برای این کار  هر یالی که بین دو مجموعه سمت راست و چپ وجود دارد را در نظر میگیریم ، یال را از راس سمت چپ به راس سمت راست جهت دار میکنیم .
دو راس sو t در نظر میگیریم . از راس s به همه رئوس مجموعه چپ یال جهت دار میکشیم و از همه رئوس مجموعه راست به راس t نیز یال جهت دار میکشیم و سپس ظرفیت همه یال ها را برابر با یک قرار میدهیم . حال برای گراف H میتوانیم مساله 
بیشینه شار را از مبدا s به مقصد t حل کنیم .
<br>
این تغییرات برای یک گراف 2 بخشی دلخواه در شکل زیر اعمال شده است .

حال ادعا میکنیم پیدا کردن ماکسیمم شار در گراف ساخته شده برابر با ماکسیمم تطابق در گراف اولیه است .
هر واحد شار را اگر در نظر بگیریم اینگونه است که مسیری از s به یکی از رئوس سمت چپ و از آن راس  به یکی از رئوس سمت راست و سپس به راس t است .
در تیجه ظرفیت همه یال هایی که طی کردیم پس از عبور یک واحد شار برابر با صفر خواهد شد . با توجه به اینکه جهت همه یال ها از سمت چپ به سمت راست است پس بعد از این هیچ گاه ظرفیت یال ها تغییر نمیکند و نتیجتا ظرفیت همین صفر میماند 
پس دیگر هیچواحد شاری وجود نخواهد داشت که به راس سمت راست یا چپ برسد زیرا یال متصل از s و t به آنهاانگار دیگر وجود ندارد 
این یک واحد شار عبور داده شده برابر با یک عضو از تطابق است زیرا باعث میشود پس از عبور شار دو راس وسطی که از آنها شار عبور داده شده حذف شوند و تمامی یال های متصل به آنها نیز در واقع حذف شوند (زیرا هیچ گاه به این دو راس نخواهم رسید )
که این مساله همان چیزی است که در مساله پیدا کردن تطابق میخواهیم .
بنابراین چون هر واحد شار یک عضو تطابق است پس ماکسیمم شار هم ماکسیم تطابق است 
و اینگونه اثبات میشود که اگر بتوانیم ماکسیمم شار را در گراف ساخته شده پیدا کنیم ، ماکسیمم تطابق هم در گراف اولیه برابر خواهد بود با یال های میانی که ظرفیتشان صفر شده است .

مسئله‌ی سوم: گرد کردن ماتریس
</div>
فرض کنید یک ماتریس $m \times n$ به نام D داریم. در این ماتریس جمع اعضای سطر $i$ را با $r_{i}$ و جمع اعضای ستون $j$ را با $c_{j}$ نمایش می‌دهیم.
هدف از این مسئله این است که همه‌ی اعضای ماتریس ($d_{ij}$) و همچنین $r_i$ و $c_i$ ها را طوری گرد کنیم که جمع‌ها پایدار باشند. به این معنا که جمع اعضای گرد شده‌ی ماتریس در سطر $i$ با $r_i$ 
گرد شده برابر باشند و این موضوع برای ستون‌ها نیز صدق کند.
<br>
شاید اولین راه که برای حل این مسئله به ذهن برسد این باشد که یک حد آستانه مانند 0.5 برای گرد کردن قرار دهیم و اعدادی که اعشار آن‌ها از 0.5 بیشتراست را به بالا و آن‌هایی که اعشارشان از 0.5 کمتر است را رو به پایین گرد کنیم اما همانطور که در مثال پایین می‌بینید این روش اشتباه است.

<figure>
    <img src="resources/matrix_rounding.png" , style="float:center">          
</figure>
 همچنین در شکل بالا ما‌توانید یکی از راه‌حل های این مثال را مشاهده کنید. برای حل این مسئله می‌توان از الگوریتم یافتن شار بیشینه استفاده کرد. اما ابتدا باید مسئله را به صورت یک گراف مدل کنیم.
 <br>
 به این منظور یک 
 گراف دوبخشی می‌سازیم که یک بخش مربوط به سطر‌ها و یک بخش مربوط به ستون‌ها است. بین هر راس مربوط به  
 سطر $i$ و یک راس مربوط به ستون
 $j$ یک  یال وجود دارد که مربوط به 
 عضو $d_{ij}$ ماتریس است.
 همچنین باید به این گراف یک راس سورس و یک راس سینک اضافه کنیم به این صورت که راس سورس (S) به همه‌ی راس‌های بخش سطر با یک یال متصل است و راس سینک (t) 
 به همه‌ی راس‌های بخش ستونی متصل است.
 <br>
 ظرفیت کمینه‌ی هر سطر مربوط به اعضای ماتریس برابر گرد شده‌‎ی آن عضو رو به پایین و ظرفیت بیشینه‌ی آن برابر با گرد شده‌ی رو به بالای آن ماتریس است.
 برای یال‌های بین سورس و سینک و راس‌های مربوط به سطر و ستون نیز کمینه‌ی ظرفیت برابر گرد شده‌ی رو به پایین مجموع آن سطر یا ستون و بیشینه‌ی ظرفیت برابر گرد شده‌ی رو به بالای آن سطر یا ستون است.
حال اگر معادله‌ی شار ورودی و خروجی برای هر راس را بنویسید، واضح است که این معادلات همان محدودیت‌هایی هستند که با حل آن‌ها مسئله‌ی گرد کردن ماتریس حل می‌شود.
<br>
گرافی که برای ماتریس مثال بالا ساخته شده است را می‌توانید در زیر مشاهده کنید.
