<div id="section1" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
مقدمه
</div><br>
همان‌طور که می‌دانید، پیدا کردن کوتاه‌ترین مسیر بین رأس‌ها در گراف‌ها، کاربرد بسیاری دارد. از جمله در پیدا کردن بهترین مسیر برای رانندگی، طراحی شبکه جاده‌ای، مسیریابی شبکه (Network Routing) , ... . در درس ساختمان داده، با الگوریتم‌های پیدا کردن کوتاه‌ترین مسیر از یک مبدأ (Single-Source Shortest Paths) از جمله بلمن-فورد و دایکسترا آشنا شدیم. این الگوریتم‌ها، به ازای هر رأس از گراف، کوتاه‌ترین مسیر برای رسیدن از رأسی مشخص به آن رأس را پیدا می‌کنند. حال اگر بخواهیم بین هر دو رأس در گراف، کوتاه‌ترین مسیر را پیدا کنیم، چه باید کرد؟    
</div>
<hr>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
سوال؛ کوتاه‌ترین مسیر بین هر دو راس
</div><br>
در یک گراف جهت‌دار و وزن‌دار بدون دور منفی، به ازای هر دو رأس $u,v \in V$، کوتاه‍ترین مسیر (مسیر با کمترین وزن) بین $u$ و $v$ را بیابید.</div>
<hr>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
راه حل اولیه</div><br>            
اولین راهی که به ذهن می‌رسد، این است که به ازای هر رأس، یک بار الگوریتم دایکسترا (Dijkstra) را به مبدأ آن رأس اجرا کنیم تا برای هر رأس، کوتاه‌ترین مسیرها به مبدأ این راس را پیدا کنیم، و به این ترتیب، کوتاه‌ترین مسیر بین هر دو رأس را پیدا کنیم. این راه‌حل به خوبی جواب می‌دهد و مرتبه زمانی آن برابر با $|V|$ بار اجرای دایکسترا است، یعنی $O(V(V\lg V + E)) = O(V^2\lg V + VE)$ است.
<br>
اما یک نکته‌ی مهم را باید در نظر گرفت؛ اگر گراف یال منفی داشته باشد، الگوریتم دایکسترا جواب درست نمی‌دهد! در این حالت چه کنیم؟
<hr>
</div>    

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">      
شاید بگویید می‌توانیم به‌جای دایکسترا، از الگوریتم بلمن-فورد (Bellman-Ford) استفاده کنیم، که در گراف‌های با یال منفی نیز پاسخ درست می‌دهد. (فقط کافی است گراف، دور منفی نداشته‌باشد.) بله درست است. این الگوریتم به درستی کار می‌کند.
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
تحلیل زمانی
</div>
مرتبه زمانی الگوریتم بلمن-فورد، از $O(VE)$ است. حال اگر به ازای هر رأس به عنوان مبدأ، این الگوریتم را اجرا کنیم، مرتبه زمانی کلی از $O(V(VE)) = O(V^2E)$ می‌شود. در بدترین حالت، تعداد یال‌های گراف از $O(V^2)$ است، پس مرتبه‌ی زمانی این روش در بدترین حالت از $O(V^4)$ است.<br>
آیا می‌توان الگوریتمی با مرتبه زمانی بهتر ارائه داد؟
<hr>
</div>

<div id="johnson_algorithm" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
الگوریتم جانسون</div><br>            
اگر بتوانیم وزن یال‌ها را به گونه‌ای تغییر دهیم که وزن همه‌ی یال‌ها مثبت شود، و در عین حال کوتاه‌ترین مسیر بین هیچ دو رأسی عوض نشود، می‌توانیم از همان راه حل $V$ بار اجرای دایکسترا استفاده کنیم. حال سوال این است که چگونه این کار را انجام دهیم؟<br>
ممکن است بگویید که همه‌ی یال‌ها را با مقداری ثابت (مثلاً اندازه‌ی منفی ترین یال گراف) جمع می‌کنیم تا وزن همه‌ی یال‌ها مثبت شود. با کمی بررسی می‌توان متوجه شد این حرکت، ممکن است کوتاه‌ترین مسیر بین بعضی رئوس را عوض کند. پس چه باید کرد؟🤔 <hr>
راهی که الگوریتم جانسون پیشنهاد می‌دهد، این است: <br>
به ازای هر رأس مانند $v \in V$، یک عدد مانند $h(v)$ نسبت می‌دهیم. حال به ازای هر یال $(u,v)\in E$ با وزن $w(u,v)$، وزن جدید $\hat{w}(u,v)$ را مطابق زیر تعریف می‌کنیم:
$$\hat{w}(u,v)=w(u,v) + h(u) - h(v)$$
ابتدا باید ثابت کنیم که این تغییر در گراف، کوتاه‌ترین مسیرها را تغییر نمی‌دهد. یک مسیر دلخواه مانند $p=\langle v_0,v_1,\dots, v_k\rangle$ در نظر بگیرید. وزن این مسیر در گراف با وزن‌های جدید برابر است با:
    <br>
$$\hat{w}(p) = \hat{w}(v_0,v_1)+\hat{w}(v_1,v_2)+...+(v_{k-1},v_k)=\sum^{k}_{i=1}{\hat{w}(v_{i-1}, v_i)}\\
= \sum^{k}_{i=1}{(w(v_{i-1}, v_i) + h(v_{i-1})-h(v_i))}\\
= \sum^{k}_{i=1}{w(v_{i-1}, v_i)} + h(v_0)-h(v_k)\\
=w(p) + h(v_0)-h(v_k)$$
    <br> 
</div>
<div id="section2" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
پس مشاهده می‌شود که مسیرهایی با ابتدا و انتهای مشخص، وزن‌شان به مقدار ثابتی ($h(v_0)-h(v_k)$) تغییر می‌کند. یعنی اگر مسیری در گراف اصلی، کوتاه‌ترین مسیر بین دو رأس $u$ و $v$ باشد، در گراف با وزن‌های جدید نیز کوتاه‌ترین مسیر بین این دو رأس است، و بالعکس. <br>
همچنین بر اثر این تغییر وزن یال‌ها، وزن دورهای گراف تغییری نمی‌کند. پس گراف با وزن یال‌های جدید، دور منفی ندارد اگر و تنها اگر گراف اصلی دور منفی نداشته باشد.
<hr>
سپس، باید یک روش مناسب برای محاسبه تابع $h$ برای رأس‌ها پییشنهاد دهیم، به طوری که وزن جدید تمام یال‌ها مثبت شود.<br>
روش الگوریتم جانسون این است؛ گراف $G'$ را از روی گراف اصلی $G$ به این شکل می‌سازیم: یک راس جدید مانند $s \notin V$ به گراف $G$ اضافه می‌کنیم، سپس از $s$ به رأس‌های دیگر، یالی با وزن 0 قرار می‌دهیم. مشخص است که این گراف $G'$ دور منفی ندارد اگر و تنها اگر $G$ دور منفی نداشته باشد. حال روی گراف $G'$ الگوریتم بلمن-فورد را با مبدأ $s$ اجرا می‌کنیم و به ازای هر رأس $v\neq s$، $h(v)$ را برابر با $d(v) = \delta(s,v)$ قرار می‌دهیم.<br>
ادعا می‌کنیم که گراف $G$ بر اثر تغییر وزن یال‌ها با استفاده از تابع $h$ مذکور، دیگر یالی با وزن منفی نخواهد داشت. <br>
در الگوریتم بلمن-فورد، پس از اتمام اجرا ($|V|-1$ بار Relax کردن تمام یال‌ها)، به ازای هر یال مانند $(u,v)\in E$ داریم: 
$$d_u+w(u,v) \ge d_v \Rightarrow w(u,v) + d_u - d_v \ge 0 \Rightarrow \hat{w}(u,v)\ge 0$$
مشخص است که وزن جدید تمام یال‌ها بر اثر این تغییر، مثبت می‌شود. پس این روش برای محاسبه تابع $h$، روش مناسبی است.
</div>  

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
تحلیل زمانی
</div>
در ابتدا، برای محاسبه‌ی مقادیر $h(v)$ به ازای رئوس، یک بار الگوریتم بلمن-فورد را روی گراف $G'$ با $|V|+1$ راس و $|E|+|V|$ یال اجرا می‌کنیم. مرتبه‌ی زمانی این مرحله برابر از $O((|V|+1)(|V|+|E|))=O(VE)$ است. (اگر $|V| = O(|E|)$ باشد. در غیر این صورت، یعنی گراف ناهمبند است، و الگوریتم بلمن-فورد را برای هر مؤلفه همبندی به صورت جدا اجرا می‌کنیم.) سپس $|V|$ بار الگوریتم دایکسترا را اجرا می‌کنیم، که اگر با استفاده از هرم فیبوناچی پیاده‌سازی شود، زمان اجرای آن از مرتبه‌ی $O(V(V\lg V+E)) = O(V^2\lg V +VE)$ است. پس زمان اجرای کلّی الگوریتم از مرتبه‌ی $O(VE+V^2\lg V + VE)=O(V^2\lg V + VE)$ است. <br>
این مرتبه زمانی در بدترین حالت (زمانی که $|E| = |V|^2$ باشد) برابر با $O(V^3)$ است، که برابر با مرتبه زمانی الگوریتم فلوید-وارشال است. اما در گراف‌های خلوت (Sparse)، زمان اجرای آن به صورت حدی از الگوریتم فلوید-وارشال کم‌تر است.
<br/>
می‌توانید کد این الگوریتم را در زیر ببینید.
<hr>
</div>

<div id="another_method" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
روشی دیگر
</div><br>
خب بیاید با ابزار دیگری به سراغ حل مسئله بیان شده برویم.
 آیا اگر کمترین فاصله دو راس دلخواه را داشته باشیم، برای پیدا کردن کمترین فاصله دو راس دیگر از این اطلاعات قبلی می‌توانیم استفاده کنیم؟
    به عبارتی دیگر مسئله را می‌توانیم به یک سری مسائل کوچکتر تبدیل کنیم  که با حل آنها مسئله اصلی نیز حل شوند؟

<hr>
همانگونه که احتمالا حدس زده‌اید، برای حل این مسئله از ایده برنامه‌ریزی پویا می‌خواهیم استفاده کنیم و می‌خواهیم به گونه‌ای فاصله میان رئوس را به یک سری مسائل کوچک دیگر که جواب آن‌ها را قبلا محاسبه کرده‌ایم تبدیل کنیم.

<hr>
فرض کنید که کم‌ترین فاصله میان دو راس 
    $i$ و $j$
    را با $d[i,j]$
    نشان بدهیم.
    حال می‌دانیم که نامساوی زیر برای هر سه راس دلخواه $i,j,k$
    برقرار می‌باشد.
    $$
    d[i,j] \le d[i,k] + d[k,j]
    $$
    بنابراین برای محاسبه
    $d[i,j]$
    می‌توان تمام رئوس $k$ 
    ممکن را در نظر گرفت و مقدار سمت راست نامساوی بالا را برای آن‌ها محاسبه کرد و خواهیم داشت:
    $$
    d[i,j] = \min_{k} \bigg\{d[i,k] + d[k,j]\bigg\}
    $$
    اگر در رابطه به دست آمده دقت کنید می‌بینید که این رابطه یک مشکل اساسی دارد. این رابطه وابستگی دوری دارد و به درستی عمل نخواهد کرد. یعنی با استفاده از آن هیچ‌گاه نخواهیم توانست کمترین فاصله میان رئوس را پیدا کنیم.
    به عبارت دیگر در برای حل یک پرسش با پرسشی از همان جنس مواجه شدیم. اما با تغییر اندکی در رابطه بالا و به گونه‌ای ایجاد ترتیب  در حل زیرمسئله‌ها می‌توان این مشکل را حل کرد.
    <hr>
</div>

<div id="floyd_algorithm" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
الگوریتم فلوید وارشال
</div><br>
در الگوریتم فلوید وارشال به جای 
    $d[i,j]$
    که کوتاه‌ترین مسیر میان $i$ و $j$ 
    می‌باشد، 
    $d[i,j,k]$ 
    را داریم که کوتاهترین مسیر میان رئوس
    $i$ و $j$
    با شرط عبور از رئوس 
    $1$ تا $k$ 
    است.
    به بیان دیگر رئوس میانی مسير صرفا رئوس 
    $1$ تا $k$
    هستند (فرض کنید رئوس گراف از 
    $1$ تا $n$ 
    شماره گذاری شده‌اند).
    با این فرض در هر مرحله می‌توان نوشت:
    $$
    d[i,j,k] = min(d[i,j,k-1], d[i,k,k-1] + d[k,j,k-1]), \hspace{1em} k \ge 1
    $$
   یعنی در کوتاهترین مسیر میان
   $i$ و $j$
   که رئوس میانی از 
   $1$ تا $k$
   باشند، یا از راس $k$ام
   استفاده شده است یا نه.
   اگر استفاده نشده باشد که که پاسخ همان 
   $d[i,j,k-1]$
   می‌باشد وگرنه که بایست کمترین مسیر از $i$ به $k$ 
   و از $k$ به $j$ 
   که صرفا از رئوس 
   $1$ تا $k-1$
   استفاده شده است را با هم جمع کنیم و مینیمم مقدار این دو حالت برابر با 
   $d[i,j,k]$
   خواهد بود.
   <br>
   پایه الگوریتم نیز برای
   $k = 0$
   خواهد بود که یعنی هیچ راس میانی‌ای نداریم و مقدار 
   $d[i,j,0]$
   اگر دو راس 
   $i$ و $j$
   مجاور باشند، وزن یال بین 
   $i$ و $j$
   (که با 
   $w[i,j]$
   نشان می‌دهیم)
   و در غیر این صورت بی‌نهایت خواهد بود.
   یعنی داریم:
   $$
   d[i,j,0] = \begin{cases}
        w[i,j]& \text{if there is edge between i and j} \\
        \infty&  \text{otherwise}\\
        \end{cases}
   $$
   <br>
   حال پاسخ سوال اولیه ما یعنی طول کوتاهترین مسیر میان دو راس دلخواه یا فاصله دو راس دلخواه چیست؟
    فاصله دو راس دلخواه 
    $i$ و $j$
    برابر با $d[i,j,n]$
    خواهد بود (یعنی تمامی رئوس 
    $1$ تا $n$
    در طول مسیر می‌توانند حضور داشته باشند).
<hr>
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
تحلیل زمانی الگوریتم فلوید وارشال
</div>
برای تحلیل زمانی الگوریتم تعداد زیرمسئله‌هایی که داریم و مقدار محاسباتی که باید برای حل هر کدام انجام بدهیم را بررسی می‌کنیم.
    در این الگوریتم ما مقدار 
    $d[i,j,k]$
    را به ازای 
    $ 1 \le i,j,k \le n$
    باید محاسبه کنیم. بنابراین تعداد مسائلی که باید حل کنیم از 
    $O(n^3)$
    خواهد بود و برای حل هر کدام از این مسائل به اندازه 
    $O(1)$
    هزینه صرف می‌کنیم.
    در نتیجه پیچیدگی زمانی الگوریتم فلوید وارشال از 
    $O(n^3)$
    خواهد بود.
<br>
<hr>
سوال: اگر با الگوریتم فلوید وارشال بخواهیم کمترین فاصله میان تمام رئوس را به دست آوریم، حافظه مورد نیاز در چه پیچیدگی‌ای خواهد بود؟
<br>
پاسخ:
با $O(n^2)$
حافظه می‌توان این الگوریتم را پیاده سازی کرد (چرا؟).
<hr>
مسئله:
یک گراف دلخواه با $n$ راس و $m$ یال داریم.
در هر مرحله با دو نوع درخواست مواجه می‌شویم.
درخواست نوع اول این است که یک راس از گراف حذف شود و درخواست نوع دوم این است که فاصله میان دو راس دلخواه در گراف باقیمانده تا آن لحظه چه قدر است.
اگر تعداد درخواست‌ها $q$ باشد.
الگوریتمی از پیچیدگی 
$O(n^3)$
برای حل مسئله گفته شده ارائه دهید ($q << n^3$).
(راهنمایی: حذف شدن یک راس را می‌توان اضافه شدن راس در نظر گرفت اگر درخواست‌ها  را 
offline
و از آخر به اول بررسی کنیم)
<br>
منبع: تمرین عملی درس طراحی الگوریتم نیمسال اول سال تحصیلی ۹۸-۹۹
</div>

```python
#include <iostream>
using namespace std;
const int MAX=1000, MAXINT = 1000000000;

void AllPairsShortestPaths(int G[MAX][MAX], int n) {
  int d[MAX][MAX];
  for (int i = 0; i<n; i++)
    for (int j = 0; j < n; j++)
      d[i][j] = G[i][j];

  for (int k = 0; k < n; k++)
    for (int i = 0; i < n; i++)
      for (int j = 0; j < n; j++)
        if (d[i][k] + d[k][j] < d[i][j])
          d[i][j] = d[i][k] + d[k][j];

  for (int i = 0; i < n; i++)
  {
    for (int j = 0; j < n; j++)
      cout << d[i][j] << "\t";
    cout << endl;
  }
}

int main(int argc, char ** argv){
  int n, m;
  cin >> n >> m;
  int G[MAX][MAX];
  for (int i = 0; i < n; i++)
    for (int j = 0; j < n; j++)
      if (i != j)
        G[i][j] = MAXINT;
      else
        G[i][j] = 0;
  for (int i = 0; i < m; i++)
  {
    int u, v, w;
    cin >> u >> v >> w;
    G[u-1][v-1] = w;
  }
  AllPairsShortestPaths(G, n);
  return 0;
}
```
