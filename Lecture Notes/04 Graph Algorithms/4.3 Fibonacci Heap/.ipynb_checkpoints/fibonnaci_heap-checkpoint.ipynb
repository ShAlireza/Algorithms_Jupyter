مقدمه
</div>
    <div>
    تعریف: فیوناچی هیپ یا هرم فیبوناچی یک داده ساختار برای پیاده سازی صف اولویت می‌باشد که و رانینگ تایم سرشکن بهتری از تمامی دیگر صف اولویت هاست.
     <img src="img/pic1.png"/>
    <div style="font-size: 16px; color: #ff6600; font-weight: bold">
        نکته:
    </div>
        <div style="background-color: lightblue;">
        در فیبوناچی هیپ بر خلاف دیگر هرم ها تحلیل امورتایزد است.
        </div>    
    </div>
    <div style="direction:rtl";>
       فیبوناچی هیپ مجموعه‌ای از درخت ها به شکل تمامی ماکس هیپ یا تمامی مین هیپ 
        است.
        نکته دیگر این است که این درخت ها می‌توانند هر شکلی داشته باشند حتی تمامی این درخت ها می‌توانند تنها متشکل از یک راس باشند.
    <br>
     هرم فیبوناچی یک پوینتر به مینیموم خود دارد(اینجا برای هرم کمینه توضیح می‌دهیم.)
     ریشه تمام این درخت ها با یک لینک لیست دو طرفه و مدور(منظور این است که آخری به اولی لینک دارد)
        به هم متصل هستند.
        (circular doubly linked list)
    </div>
    <img src="img/pic3.png"/>
    <div style="direction:rtl">
        استفاده از لینک لیست دوطرفه و مدور دو مزیت دارد یک اینکه اضافه کردن یک عضو در O(1) انجام می‌شود مورد دوم نیز این است که اگر دو لیست به ما داده شود می‌توان آنها را در O(1) ترکیب کرد.   
    </div>
    <div style="direction:rtl;">
         ایده اصلی در هرم فیبوناچی این است که همه عملیات را به شکل «lazy»
        انجام دهیم، به این معنی که مثلا عملیات مرج خیلی ساده دو درخت را به هم لینک می‌کند و عملیات 
        اضافه کردن یک راس نیز به سادگی یک درخت تک راس را به مجموعه درخت ها اضافه می‌کنیم.
    </div>
    <div style="direction:rtl;">
    یادآوری:الگوریتم دایسترا (dijkstra)
    <br>
     همانطور که میدانید این الگوریتم به ازا یک سورس کوتاه‌ترین مسیر را به دیگر راس‌ها به ما می‌دهد اگر یال‌ها وزن منفی نداشته باشند.
     <br>
        در هر مرحله این الگوریتم باید دو تا کار را انجام دهید:
    <br>
    <ol>
        <li>راس v با کمترین فاصله از S را پیدا کنید</li>
        <li> فاصله‌های کاندید شده برای همه راس ها متصل به V را(همسایه های آن)
        آپدیت کنید(فاصله فقط در این مرحله آپدیت می‌شود.)</li>
    </ol>
    اولین مرحله مشابه extract-min 
     کردن در یک صف اولویت است اما برای مرحله دوم چه کنیم؟
    اینجا مفهوم decrease-key وسط می‌آید که کمی مفصل‌تر آنرا بررسی خواهیم کرد.
    </div>
    <div style="direction:rtl;" >
    یادآوری۲:الگوریتم پرایم:
        <br>
    این الگوریتم درخت پوشا کمینه در یک گراف بدون جهت را پیدا می‌کند.
    <br>
    در هر مرحله از این الگوریتم:
    <ol>
        <li>راس v را پیدا کنید به طوری که اولا (تا این مرحله)
        خارج از درخت ما باشد و کمترین هزینه را برای اضافه شدن داشته باشد.
        </li>
        <li>فاصله کاندید شده برای V به راس های خارج از ست کنونی درخت </li>
        را بروزرسانی کنید(مشابه مرحله دوم دایسترا)
        </ol>
        بار دیگر مشخص است مرحله اول همان extract-min و مرحله دوم decrease-key خواهد بود.
    </div>

        اجتماع و اضافه کردن
    </div>
    <div style="font-size: 16px; color: #ff6600; font-weight: bold">
        اضافه کردن
        (Insertion)
        :
    </div>
    <div>
        برای اضافه کردن یک راٰس مراحل زیر را به ترتیب انجام می‌دهیم:
        <ol>
            <li>راس V را می‌سازیم</li>
            <li>چک می‌کنیم که آیا هرم ما خالی است یا خیر</li>
            <li>اگر خالی بود V را تنها راس در مجموعه راس های root list قرار می‌دهیم 
            و V را برابر مینیموم هرم می‌گذاریم
                (root list همان لینک لیست دوطرفه برای نگه داشتن ریشه درخت ها است که در بخش پیش معرفی شد)
            </li>
            <li>در غیر این صورت  V را به root list اضافه می‌کنیم و در صورت نیاز مینیموم را آپدیت می‌کنیم.</li>
        </ol>
        مثالی از این مرحله:
        <br>
        <img src="img/pic4.jpg"/>
    </div>
    <div style="font-size: 16px; color: #ff6600; font-weight: bold">
        اجتماع
        (Union)
        :
    </div>
    <div style="direction:rtl;">
        برای اجتماع دو فیبوناچی هیپ مراحل زیر را اجرا می کنیم:
        <ol>
            <li>ریشه های دو هرم را در کنار هم در یک root list قرار می‌دهیم</li>
            <li>if H1(min) &lt H2(min) then H(min) = H1(min) </li>
            <li>else: H(min) = H2(min)</li>
        </ol>
        H(min) = مینیموم هرم حاصل
        <br>
        H1(min) = مینیموم هرم اول
        <br>
        H2(min) = مینیموم هرم دوم
    </div>

        حذف ، کاهش کلید و استخراج کمینه
    </div>
    <div style="font-size: 22px; color: #cf3abf; font-weight: bold">
    بخش اول:
    استخراج کمینه
    O(log n)
    </div>
    <div>
         برای اینکار یک عملگر ترکیب کردن
        دو درخت را معرفی می‌کنیم
       (Merge):
        <br>
        دو درخت با رنک برابر k را به شیوه زیر ترکیب می‌کنیم:
        <ol>
            <li>آن درختی که ریشه بزرگتر را دارد فرزند دیگری قرار می‌دهیم</li>
            <li>درخت نهایی بدست آمده دارای رنک k+1 خواهد بود.</li>
        </ol> 
        <img src="img/pic6.png" />
        <div style="font-size: 16px; color: #ff6600; font-weight: bold">
        نکته:
    </div>
        <div style="background-color: lightblue;">
           Merge که در اینجا گفته شد همان Union مرحله پیشین است اما در عمل فقط برای دو درخت با رنک برابر در هرم فیبوناچی این کار را انجام می‌شود.     
        </div>  
        حال در استخراج کمینه از تعریف بالا بهره می‌بریم، اما
        به طور کلی این عمل از مراحل زیر تشکیل شده است.
        <ol>
            <ul>
                <li>گره کمینه را حذف می‌کنیم</li>
                <li>مینیموم را بروزرسانی می‌کنیم و تمامی زیر درخت های گره حذف شده را به root list اضافه می‌کنیم</li>
            </ul>
            <img src="img/pic9.jpg" />
                <li>یکی کردن درخت ها به طوری که در نهایت هیچ دو درختی دارای رنک یکسان نباشند.(برای این مرحله از هما تعریف ترکیب 
           که در پاراگراف بالا انجام شد بهره می‌بریم به این نحو که اینقدر این کار را تکرار می‌کنیم تا هیچ دو درختی دارای رنک یکسان نباشند.)
            </li>   
        </ol>
        به مثال زیر دقت کنید که پس از تکرار چندباره مرحله سوم چگونه هرم ما در رنک زیر درخت ها یکتا می‌شود:
        <img src="img/pic10.jpg"/>
    </div>
    <div style="font-size: 22px; color: #cf3abf; font-weight: bold">
    بخش دوم:
    کاهش کلید
    O(1)
    </div>
    <div>
        برای این مرحله ابتدا یک فیلد جدید در هر گره تعریف می‌کنیم که مشخص می‌کند آیا یک گره تا کنون علامت خورده است یا خیر حال الگوریتم را بیان می‌کنیم و از آن برای اینکه عمل کاهش کلید ما در O(1) انجام شود بهره می‌بریم:
        <ul>
            <li>مقدار گره V را به مقدار گفته شده کاهش می‌دهیم و سه اتفاق ممکن است رخ دهد:</li>
            <li>حالت اول:</li>
            <ul>
                <li>اگر خواص هرم کمینه منقضی نشد(از پدرش کمتر نشد)</li>
                <li>تنها در صورت نیاز مینیموم را بروزرسانی می‌کنیم</li>
            </ul>
            <li> حالت دوم: اگر خواص هرم کمینه نقض شد و پدرش علامت نخورده بود</li>
            <ul>
                <li>ارتباط بین V و پدرش را قطع می‌کنیم</li>
                <li>پدرش را علامت می‌زنیم</li>
                <li>درختی که ریشه‌اش V باشد را به root list اضافه می‌کنیم</li>
            </ul>
            <li> حالت سوم:اگر هم خواص هرم کمینه نقض شده باشد هم پدرش علامت خورده باشد</li>
                <ol>
                    <li>ارتباط بین V و پدرش را قطع می‌کنیم</li>
                    <li>درختی که ریشه آن V است را به root list اضافه می‌کنیم و در صورت نیاز کمینه را بروزرسانی می‌کنیم</li>
                    <li>ارتباط بین p[V] و p[p[V]] را قطع می‌کنیم</li>
                    <li>درختی که ریشه اش p[V] می‌باشد را به root list اضافه می‌کنیم(این درخت دیگر شاخه V را ذر خود ندارد)</li>
                    <li> در صورت نیاز کمینه را بروزرسانی می‌کنیم</li>
                    <li>اگر p[p[V]] علامت زده نشده بود آنرا علامت می‌زنیم</li>
                    <li>در غیر این صورت اورا نیز جدا کرده و مراحل ۲ تا ۶ را با p[p[V]] تکرار می‌کنیم</li>
                </ol>
        </ul>

    بخش سوم:
    حذف
    O(log n)
    </div>
    با توجه به ابزارهایی که در بالا تغریف شد این مرحله بسیار ساده
   در O(log n)
    انجام پذیر است.
    <div>
        برای دیلیت کردن یک گره مراحل زیر را انجام می‌دهیم:
        <ul>
            <li>مقدار گره‌ای که قرار است حذف شود را به کمک Decrease_Key به مقداد کمینه کاهش می‌دهیم</li>
            <li>مانند بخش قبل دوباره درختی که ریشه آن V  است را به root list اضافه می کنیم</li>
            <li>حال الگوریتم Extract_Min که در بخش اول توضیخ داده شد را انجام می‌دهیم</li>
        </ul>
        <div style="font-size: 16px; color: #ff6600; font-weight: bold">
        نکته:
    </div>
        <div style="background-color: lightblue;">
           چون در این الگوریتم از Extract_min استفاده می‌کنیم O(log n) می‌شود.
