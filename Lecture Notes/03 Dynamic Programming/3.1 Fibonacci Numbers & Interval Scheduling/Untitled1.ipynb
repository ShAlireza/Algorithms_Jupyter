```python
#include <iostream>
#include <string>
using namespace std;
const int MAX = 1000;


int EditDistance(string A, string B) {
    int D[MAX+1][MAX+1];
    int F[MAX+1][MAX+1];
    int m = A.size();
    int n = B.size();
    for (int i = 0; i <= m; i++) {
        D[i][0] = i;
        F[i][0] = 3;
    }
    for (int j = 0; j <= n; j++) {
        D[0][j] = j;
        F[0][j] = 2;
    }

    for (int i = 1; i <= m; i++)
        for (int j = 1; j <= n; j++) {
            if (A[i-1] == B[j-1]) {
                D[i][j] = D[i-1][j-1]; // Match
                F[i][j] = 0;
            }
            else {
                D[i][j] = D[i-1][j-1] + 1; // Mutation
                F[i][j] = 1;
            }

            if (D[i-1][j] + 1 < D[i][j]) {
                D[i][j] = D[i-1][j] + 1; // Deletion
                F[i][j] = 2;
            }

            if (D[i][j-1] + 1 < D[i][j]) { // Insertion 
                D[i][j] = D[i][j-1] + 1;
                F[i][j] = 3;
            }
        }
    
    int i = m, j = n;
    while (i > 0 || j > 0) {
        switch (F[i][j]) {
            case 0:
            cout << "Match "  << i << " " << j << endl;
            i--;
            j--;
            break;
            
            case 1:
            cout << "Mismatch "  << i << " " << j << endl;
            i--;
            j--;
            break;
           
            case 2:
            cout << "Delete "  << i << " " << j << endl;
            i--;
            break;
            
            case 3:
            cout << "Insert "  << i << " " << j << endl;
            j--;
            break;       
        }       
    }
    
    for (int i = 0; i <= m; i++) {
        for (int j = 0; j <= n; j++) 
            cout << D[i][j] << '\t';
        cout << endl;
    }
    return D[m][n];
}

int main(int argc, char * argv[]) {
    string A, B;
    cin >> A;
    cin >> B;
    cout << EditDistance(A, B) << endl;
    return 0;
}
```
